## Ch7 The Capital Asset Pricing Model

### **Market Equilibrium**
From the one-fund theorem every individual will form a portfolio that is a mix of the risk-free asset and the single, risky *one* fund.<br/>Hence, the followings can be derived :
1. the one fund in the theorem is really the *only* fund that is used and must equal the *market portfolio*
    - Market Portfolio : portfolio with weights equal to the proportion of that assetâ€™s total capital value to the total market capital value(*capitalization weights*)
    - Equilibrium Argument<Br/>: investors solve the mean-variance portfolio problem using their common estimates, and they place orders in the market to acquire their portfolios. If the orders placed do not match what is available, the prices must change. The prices of assets under heavy demand will increase; the prices of assets under light demand will decrease. In other words, prices adjust to drive the market to efficiency. These price changes affect the estimates of asset returns directly, and hence investors will recalculate their optimal portfolios. This process continues until demand exactly matches supply; that is, it continues until there is equilibrium. After other people have made the adjustments, we can be sure that the efficient portfolio is the market portfolio

2. the efficient set therefore *consists of a single straight line, CML* on the $\bar r-\sigma$ diagram 
    - The Capital Market Line(CML) : straight line, emanating from the riskfree point and passing through the market portfolio $$E(r_p) = r_f + \frac{E(r_m) - r_f}{\sigma_m} \cdot \sigma_p$$ The slope is $K = \frac{F_y - r_y}{\omega}$, implying the price of risk

        

### **Capital Asset Pricing Model(CAPM)**

*If the market portfolio $M$ is efficient, the expected return $\overline{r_i}$ of any asset $i$ satisfies* $$\overline{r_i} - r_M = \beta_i(\overline{r_M} - r_f)$$ *where $\beta_i$ is the asset's beta, given by* $$\beta_i = \frac{\text{Cov}(r_i,r_M)}{\text{Var}(r_M)}$$ 
$\beta_i$ is referred to as the beta of an asset and $(r_i - r_f)$ is termed the expected excess rate of return of asset $i$<br/><br/>
pf) <br/>For any $\alpha$, consider the portfolio consisting of a portion $\alpha$ invested in asset $i$ and a portion $1-\alpha$ invested in the market portfolio $M$. The expected rate of return of this portfolio is
$$\overline{r_\alpha} = \alpha\overline{r_i} + (1-\alpha)\overline{r_M}$$
and the standard deviation of the rate of return is
$$\sigma_\alpha = [\alpha^2\sigma_i^2 + 2\alpha(1-\alpha)\text{Cov}(r_i,r_M) + (1-\alpha)^2\sigma_M^2]^{1/2}$$

As $\alpha$ varies, these values trace out a curve in the $\bar r$-$\sigma$ diagram, as shown in Figure 7.2. In particular, $\alpha=0$ corresponds to the market portfolio $M$. This curve cannot cross the capital market line. If it did, the portfolio corresponding to a point above the capital market line would violate the very definition of the capital market line as being the efficient boundary of the feasible set. Hence, as $\alpha$ passes through zero, the curve must be tangent to the capital market line at $M$.



The tangency condition can be translated into the condition that the slope of the curve is equal to the slope of the capital market line at the point $M$. To set up this condition we need to calculate a few derivatives. $${d\bar r_\alpha\over d\alpha}=\bar r_i-\bar r_M$$ $${d\sigma_\alpha\over d\alpha}={\alpha\sigma_i^2+(1-2\alpha)\sigma_{iM}+(\alpha-1)\sigma_M^2\over\sigma_\alpha}$$ and $${d\sigma_\alpha\over d\alpha}\bigg\rvert_{\alpha=0}={\sigma_{iM}-\sigma_M^2\over\sigma_M}$$ thus, $${d\bar r_\alpha\over d\sigma_\alpha}\bigg\rvert_{\alpha=0}={(\bar r_i-\bar r_M)\sigma_M\over\sigma_{iM}-\sigma_M^2}$$ and this slope must equal the slope of the capital market line. Hence, $${(\bar r_i-\bar r_M)\sigma_M\over \sigma_{iM}-\sigma_M^2}={\bar r_M-r_f\over \sigma_M}$$ solving for $\bar r_i$, obtaining the final result $$\bar r_i=r_f+\left({\bar r_M-r_f\over \sigma_M^2}\right)\sigma_{iM}=r_f+\beta_i(\bar r_M-r_f)$$
- expected rate of return increases linearly as the covariance with the market increases or, equivalently, as $\beta$ increases
- CAPM formula can be expressed in graphical form by regarding the formula as a linear relationship called *Security Market Line(SML)*

### **Systematic Risk of CAPM**
we can write the random rate of return of asset i as

$$
r= r_e + \beta_i(r_m-r_f) + \epsilon_i
$$

The random variable $\epsilon_i$ is chosen to make it true.Then, the CAPM says that
- taking the expected value of the above equation,  $E(\epsilon_i) = 0$
- taking the correlation of the above equation with $r_j$, $\text{cov}(\epsilon_i,r_j) = 0$ 

Thus, 
$$
\sigma_i^2 = \beta_i^2 \sigma_m^2 + \text{var}(\epsilon_i)
$$

- $\beta_i^2 \sigma_m^2$ is termed the *systematic risk*
    - risk associated with the market as a whole
    - cannot be reduced by diversification because every asset with nonzero beta contains this risk
- $\text{var}(\epsilon_i)$ is termed the *nonsystematic, idiosyncratic, or specific risk*
    - uncorrelated with the market
    - can be reduced by diversification
    - *measured by beta* which directly combines with the systematic risk of other assets

### **CAPM as a Pricing Formula**
CAPM is a *pricing model.*<br/>
*The price $P$ of an asset with payoff $Q$ is $$P = \frac{\bar Q}{1 + r_f+\beta(\bar r_M - r_f)}$$* pf)<br/> Suppose that an asset is purchased at price $P$ and later sold at price $Q$. <br/>The rate of return is then $r = (Q - P)/P$. Putting this in the CAPM formula, we have $$E(r) = r_f + \beta(E(r_m) - r_f)$$ Solving for $P$ and we can obtain the above expression.

### **Linearity of Pricing**
Suppose that we have an asset with price $P$ and final value $Q$. Here again $P$ is known and $Q$ is uncertain.<br/> Using the fact that $r= Q/P - 1$, the value of beta is $$\beta= \frac{cov[(Q/P - 1),r_M]}{\sigma_M^2}$$ Since constant additive terms do not affect covariance, this becomes $$\beta= \frac{cov(Q,r_M)}{\sigma_M^2P}$$ Thus, $$1=\frac{\bar Q}{P(1+r_f)+cov(r_M,Q)(\bar r_M-r_f)/\sigma_M^2}$$ Finally, solving for $P$ we obtain *Certainty Equivalent Form* which shows clearly that the pricing formula is linear because both terms in the brackets depend linearly on $Q$.
- Certainty Equivalent Form : The form of the CAPM pricing formula that clearly displays linearity<br/>
    The price $P$ of an asset with payoff $Q$ is $$P=\frac{1}{1+r_f}\left[{\bar Q-{cov(Q,r_M)(\bar r_M-r_f)\over \sigma_M^2}}\right]$$ 
    using $M$ for the final market value, $P$ for its price, and $R$ for the total return of the risk-free asset. In these terms, the certainty equivalent price of an asset is given by $$P=\frac{1}{R}\left[\bar Q-\frac{cov(Q,M)(\bar M-P_MR)}{\sigma_M^2}\right]$$ where now it is understood that $\sigma^2$ is the variance of $M$ (not of $R$)

- firm criterion : maximize NPV

### **Three logical choices for $M$**
Suppose there are $n$ marketed securities determined by their prices $P_1, P_2, ..., P_n$, and their payoffs $y_1, y_2, ..., y_n$, which are random.<br/>Consider the $n$ payoffs as generating a market space $\mathcal{M}$ consisting of arbitrary linear combinations of the $n$ basic payoff variables.<br/>The inner product of two random elements y and y2 is defined as $$E(y_1 y_2)$$ and we say two elements y and $y_2$ are orthogonal if $$E(y, y_2) = 0$$ Also, the norm of y is $$\|y\|=E(y^2)$$
All of these notions can be applied to elements of $M$ and the space one dimension larger, which includes $x$.

- Projection Pricing<br/>
    - Projection Pricerelation : The CAPM price $P_x$ of an asset with payoff $x$ is the price of the projection $\pi$ of $x$ in the space of marketed asset payoffs
    - CAPM pricing is equivalent to projection pricing
    - choicing M by the *original Markowitz efficient portfolio*, which is used in CAPM

        

- Minimum Norm Pricing<br/>
    Consider the set of assets in $\mathcal{M}$ that have price 1. This will be linear surface. Define the element $m_1$, as the payoff in this set of minimum norm; that is, $m_1$, is closest to the zero payoff. Now, the projection theorem, viewed from the origin, says that the inner product $E[M_1y_1]$ will be constant for all yâ€™s with price 1 because of the orthogonality relation. We scale $m_1$, to, say, $M_1$, such that for some (or any) y; with price 1 there holds $P_y=E[M_1y]$. Than even for $x$ outside $\mathcal{M}$, $$P_x=E[M_1x]$$
    - choicing M by the *minimum-norm portfolio*, which always exists

        

- Correlation Pricing<br/>
    Consider a particular asset $m \in \mathcal{M}$ is most correlated with $x$. Specifically, $m \in \mathcal{M}$ is a solution to $$\text{maximize cov}(x,m) \quad \text{subject to var}(m) \leq s^2$$
    - Correlation Pricing Formula : Let M be a marketed asset most correlated with x. Then the projection price of x may be computed as $$P_x={1\over R}\{\bar x-\beta_{x,M}[\bar M-P_MR]\}$$ where $$\beta_{x,M}=cov(x,M)/\sigma_M^2$$
    - choicing M by an asset *most correlated* with the payoff $x$

## Ch8. Other Pricing Models

### **Factor Models**
- Factors : underlying basic sources of randomness that influence the individual returns
    - must be chosen carefully : proper choice depends on the universe of assets being considered
    - Selection of Factors mgiht be..
        - real estate parcels within a city : population, employment rate, and school budgets
        - common stocks listed on an exchange : stock market average, gross national product, employment rate
        - etc : somewhat of an art, or a trial-and-error process, although formal analysis methods
    - Various factors
        1. External factors : variable that are external to the securities being explicitly considered in the model
            - GNP, CPI, unemployment rate, or a new construction index
            - it is possible to use other external variables as well, such as the number of traffic accidents in a month or sun spot activity
        2. Extracted factors : extracted factors from the known information about security returns
            - rate of return on the market portfolio(constructed directly from the returns of the individual securities)
            - the rate of return of one security can be used as a factor for others
            - an average of the returns, industrial factor, utilities factor, or transportation factor of the securities *in an industry*
            - Factors can also be extracted by the method of principal components using the covariance matrix of the returns to find combinations of securities that have large variances or in more complex ways like the ratio of the returns of two stocks, the number of days since the last market peak, or a moving average of the market return
        3. Firm characteristics : firms are characterized financially by a number of firm-specific values
            - price-earnings ratio, the dividend-payout ratio, an earnings forecast
            - The characteristics do not serve as factors in the usual sense, but they play a similar role

### **Single-Factor Model**
*assume that there is a single factor $f$ which is a random quantity and the rates of return and the factor are related by the following equation :* $$r_i = \alpha_i + \beta_i f + e_i; \qquad \text{for}\ i = 1, 2, \ldots, n.$$*$\alpha_i$'s and the $\beta_i$'s are fixed constants and $e_i$'s are random quantities which represent errors*
- errors each have zero mean(since any nonzero mean could be transferred to $\alpha_i$)$$E(e_i) = 0$$
- idealizing assumptions :
    - usually assumed that the errors are uncorrelated with $f$ and with each other; $$E[(f - \overline{f})e_i] = 0$$ for each $i$ and $E(e_i e_j) = 0$ for $i \neq j$
    - assumed that variances of the $e_i$'s are known, and they are denoted by $\sigma_i^2$
    - these assumptions may not actually be true, but are usually assumed to be true for purposes of analysis
- $a_i$ and $b_i$ are obtained for each asset i
    - since when applied to a group of assets, the fitting process is carried out for each asset separately
    - $a_i$ s are termed intercepts because $a_i$ is the intercept of the line for asset i with the vertical axis
    - $b_i$ s are termed factor loadings because they measure the sensitivity of the return to the factor

From Single-Factor Model, followings can be obtained : $$E(r_i) = a_i + b_i f$$ $$\sigma^2_i = \sigma^2_{\epsilon_i}$$ $$\sigma_{ij} = b_i b_j \sigma^2_{\epsilon_i}$$ $$b_i = \frac{\text{cov}(r_i, f)}{\sigma^2_f}$$
Suppose that a portfolio is constructed with weights $w_i$, where $\sum w_i = 1$. Then the rate of return $r$ of the portfolio is $$r=\sum_i w_ia_i + \sum_i w_ib_if + w_ie_i,$$ where $$a=\sum_i w_ia_i,$$ $$b=\sum_i w_ib_i,$$ and $$e=\sum_i w_ie_i$$
variance of $e$ is $$Var(e) = E(e^2) = E \left(\sum_i w_ie_i \right)^2 = \sum_i w_i^2 E(e_i^2) = \sum_i w_i^2 Var(e_i),$$ where we have used the fact that the $e_i$'s are uncorrelated with each other
- There are two sources of risk: that due to the $b_if$ term and that due to $e_i$
    - $e_i$ is diversifiable because it is due to the factor that influences every asset, so diversification cannot eliminate it
    - $b_i$ term is systematic or nondiversifiable risk, since they are independent and, hence, each can be reduced by diversification
- CAPM can be derived as a special case of a single-factor model<br/>
    Let us hypothesize a single-factor model for stock returns, with the factor being the market rate of return $r_M$. For convenience we can subtract the constant $r_M$ from this factor and also from the rate of return $r_i$, thereby expressing the model in terms of the excess returns $r_i-r_f$ and $r_M-r_f$. The factor model then becomes $$r_i-r_f=\alpha_i+ \beta_i(r_M-r_f)+e_i$$ The expected value of this equation is $$\bar r_i-r_f = \alpha_i + \beta_i( \bar r_m-r_f)$$ which is identical to the CAPM except for the presence of $\alpha_i$ which implies that the CAPM predicts that $\alpha_i = 0$.<br/> Form the equation, the value of $\beta_i$ is $$\beta_i={ \sigma_{iM} \over\sigma_M^2}$$ This is exactly the same expression that holds for the $\beta_i$ used in the CAPM.
    - Characteristic Line : the line formed by putting $e_i = 0$; that is, it is the line $r_i - r_f = \alpha_i + \beta_i(r_M-r_f)$ drawn on a diagram of $r_i$ versus $r_M$
        - The characteristic line is in a sense more *general* than the CAPM because it allows $a_i$ to be nonzero. From the CAPM viewpoint, $a_i$ can be regarded as a measure of the amount that asset i is *mispriced*
        - A stock with a positive $\alpha_i$ is performing better than it should, and a stock with a negative $\alpha_i$ is performing worse than it should
        - Some financial services organizations(and some highly technical investors) estimate $\alpha$ as well as $\beta$ for a large assortment of stocks
        

## **Multifactor Models**
The preceding development can be extended to include more than one factor. For example, if there are two factors $f_1$, and $f_2$, 
$$r_i= a_i + b_{1i}f_1 + b_{2i}f_2 + e_i$$

- $a_i$ is called the intercept, and $b_{1i}$ and $b_{2i}$ are the factor loadings
- the factors $f_{1}$ and $f_{2}$, and the error $e_i$ are random variables
- it is assumed that the expected value of the error is zero, and that the error is uncorrelated with the two factors and with the errors of other assets
- it is *not* assumed that the two factors are uncorrelated with each other. These factors are presumably observable variables, and their statistical properties can be studied independently of the asset returns.
- expected rates of return and the covariances $$\begin{aligned}\bar r_i &= a_i + b_{1i}\bar f_1 + b_{2i}\bar f_2\end{aligned}$$ $$\text{cov}(r_i, r_j)= \begin{cases} b_{1i}b_{1j}\sigma_{f1}^2+(b_{1i}b_{2j}+b_{2i}b_{1j})cov(f_1,f_2)+b_{2i}b_{2j}\sigma_{f2}^2, & i\neq j \\ b_{1i}^2\sigma_{f1}^2+2b_{1i}b_{2i}cov(f_1,f_2)+b_{2i}^2\sigma_{f2}^2+\sigma^2_{e_i}, & i=j \end{cases}$$
- The $b_{1i}$'s and $b_{2i}$'s can be obtained by forming the covariance of $r_i$ with $f_1$ and $f_2$, leading to $$\begin{aligned}
\text{cov}(r_i, f_1) &= b_{1i}\sigma_{f1}^2 + b_{2i}\sigma_{f1,f2} \\
\text{cov}(r_i, f_2) &= b_{1i}\sigma_{f1,f2} + b_{2i}\sigma_{f2}^2
\end{aligned}$$ These give two equations that can be solved for the two unknowns $b_{1i}$ and $b_{2i}$



## **The Fama-French (FF) Three-Factor Model**
$$R_{it} = \alpha_i + \beta_{iM}R_{Mt} + \beta_{iSMB}SMB_t + \beta_{iHML} HML_t + e_{it}$$
where
- SMB = Small Minus Big (i.e., the return of a portfolio of small stocks in excess of the return on a portfolio of large stocks)
- HML = High Minus Low (i.e., the return of a portfolio of stocks with a high book-to-market ratio in excess of the return on a portfolio of stocks with a low book-to-market ratio)
- While SMB and HML are not themselves obvious candidates for relevant risk factors, the argument is that these variables may proxy for hard-to-measure more-fundamental variables
- extra-market factors in these models cannot be clearly identified with a source of risk that is of obvious concern to a significant group of investors
- Black points out that when researchers scan and rescan the database of security returns in search of explanatory factors (an activity often called data-snooping), they may eventually uncover past â€œpatternsâ€ that are due purely to chance*(Fischer Black, â€œBeta and Return,â€ Journal of Portfolio Management 20 (1993), pp. 8â€“18.)*
- *Eugene F. Fama and Kenneth R. French, â€œMultifactor Explanations of Asset Pricing Anomalies,â€ Journal of Finance 51 (1996), pp. 55â€“84.*


## **Arbitrage Pricing Theory**
- Key propositions
    1. Security returns can be described by a factor model
    2. there are sufficient securities to diversify away idiosyncratic risk
    3. well-functioning security markets do not allow arbitrage opportunities to persist
- The Law of One Price :if two assets are equivalent in all economically relevant respects, then they should have the same market price
- risk arbitrage : mispriced securities in specific areas such as merger-target stocks, rather than strict (risk-free)arbitrage opportunities

Assume that all asset rates of return satisfy the following one-factor model: $$r_i= a_i+b_if$$

- No error term 
- The uncertainty associated with a return is due only to the uncertainty in the factor $f$ 
- values of $a_i$ and $b_i$ must be related if arbitrage opportunities are to be excluded

For two securities that $b_i \neq b_j$, form a portfolio with weights $w_i = w$ and $w_j = 1 - w$ than, the rate of return of this portfolio is $$r = w a_i + (1-w)a_j + [w b_i + + (1-w) b_j]f$$

We shall select $w$ so that the coefficient of $f$ in this equation is zero. Specifically, we select $w = \frac{b_j}{b_i - b_j}$. This yields a rate of return of

$$r=w\alpha_i+ (1-w)\alpha_j=\frac{a_ib_j}{b_j- b_i}+\frac{a_jb_i}{b_i - b_j}$$

- This special portfolio is risk-free because the equation for $r$ contains no random element
- If there is a separate risk-free asset with rate of return $r_f$, the portfolio must have this same rate â€” otherwise there would be an arbitrage opportunity
- Even if there is no explicit risk-free asset, all portfolios constructed this way, with no dependence on $f$, must have the same rate of return

We denote this common rate by $\lambda_0$, obtaining

$$\lambda_0 (b_j - b_i) = a_i b_j - a_j b_i$$

Thus,

$${a_j-\lambda_0 \over b_j}={a_i-\lambda_0 \over b_i}=c$$
$$a_i = \lambda_0 + b_jc$$

holds for all $i$ for some constant $c$ 

We can use this information to write a simple formula for the expected rate of return of asset $i$. We have

$$\bar r_i = a_i + b_i \bar f = \lambda_0 + b_i c + b_i \bar f$$

or, alternatively,

$$\bar r_i = \lambda_0 + b_i \lambda_1$$

for the constant $\lambda_1 = c + \bar f$

- if the factor $f$ is chosen to be the rate of return on the market $r_M$, and $\lambda_0 = r_f$ and $\lambda_1 =\bar r_M-r_f$, *APT is identical to the CAPM with $b_i=\beta_i$*
- once the constants $\lambda_0$ and $\lambda_i$ are known, the expected return of an asset is determined entirely by the factor loading $b_i$ (since $\alpha_i$ must follow $b_i$)
- this assumes that the universe of assets being considered is large : there are an infinite number of securities

We now consider more realistic factor models, which have error terms as well as factor terms. 

Suppose there are a total of n assets and the rate of return on asset i satisfies
$$r_i=a_i+\sum_{j=1}^m b_{ij}f_j+e_i$$
where $E(e_i)=0, E(e_i)^2=\sigma_{e_i}^2$ and $e_i$ is uncorrelated with the factors and with the error terms of other assets

$$r_j = \lambda_j + \sum_{i=1}^{m} b_{ij}f_i + e_j$$

Let us form a portfolio using the weights $w_1, w_2, ..., w_n$ with $\sum_{i=1}^{n} w_i = 1$. The rate of return of the portfolio is

$$r = a+\sum_{j=1}^{m} b_j f_j + e $$

where

$$a = \sum_{i=1}^{m} w_i a_i$$
$$b_j = \sum_{i=1}^{m} w_i b_{ij}$$
$$\sigma_e^2=\sum_{i=1}^nw_i^2\sigma_{e_i}^2$$


## **Well-Diversified Portfolio**
*Suppose that for each $i$ there holds $\sigma_{e_i}^2 < S^2$ for some constant $S$. Well diversified portfolio is that for each $i$ there holds $w_i < \frac{W}{n}$ for some constant $W \approx 1$*

- It can be obtained that $$\sigma_e^2 \leq {1\over n^2} \sum_{i=1}^{n} w^2S^2 < {1\over n}W^2S^2$$ as $n \rightarrow \infty$, $\sigma_e^2 \rightarrow 0$

## **Simple APT**

Suppose that there are n assets whose rates of return are governed by m < n factors according to the equation:

$$r_i = \sum_{j=1}^m b_{ij} f_j + a_i, for \ i = 1,2,...,n$$

Then there are constants $\lambda_0,\lambda_1,\cdots,\lambda_m$ such that:

$$\bar r_i = \sum_{j=1}^m b_{ij} \lambda_j + \lambda_0, \ for \ i=1,2,...,n$$

## **General APT**
We therefore can apply the simple APT to conclude that there are constants $\lambda_1, \lambda_2, \ldots, \lambda_m$ such that for any well-diversified portfolio having a rate of return
$$
r=a+\sum_{j=1}^m b_j f_j
$$
the expected rate of return is
$$
\bar r=\lambda_0+\sum_{j=1}^m b_j \lambda_j
$$

Since various well-diversified portfolios can be formed with weights that differ on only a small number of basic assets, it follows that these individual assets must also satisfy
$$
\bar r_i=\lambda_0+ \sum_{j=1}^m b_{ij}\lambda_j
$$
(This argument is not completely rigorous; but a more rigorous argument is quite complex)

This is again basically a relation that says that $\alpha_i$ is not independent of the $b_{ij}$'s. The risk-free term must be related to the factor loadings. This is true even when there are error terms, provided there is a large number of assets so that error terms can be effectively diversified away.

## **APT and CAPM**
The factor model underlying APT can be applied to the CAPM framework to derive a relation between the two theories.

Using a two-factor model we have
$$
r_i= a_i + b_{i1} f_1 + b_{i2} f_2 + e_i
$$
We find the covariance of this asset with the market portfolio to be
$$
\text{cov}(r_M, r_i) = b_{i1}\text{cov}(f_1, r_M) + b_{i2}\text{cov}(f_2, r_M) + \text{cov}(e_i, r_M)
$$
If the market represents a well-diversified portfolio, it will contain essentially no error term, and hence it is reasonable to ignore the term $\text{cov}(e_i, r_M)$ in the foregoing expression. We can then write the beta of the asset as
$$
\beta_i = b_{i1} \beta_1 + b_{i2} \beta_2
$$
where
$$
\begin{aligned}
\beta_{f1} &= {\sigma_{M,f_1}\over \sigma_M^2} \\
\beta_{f2} &= {\sigma_{M,f_2}\over \sigma_M^2}
\end{aligned}
$$
Hence the overall beta of the asset can be considered to be made up from underlying factor betas that do not depend on the particular asset. The weight of these factor betas in the overall asset beta is equal to the factor loadings. Hence in this framework, the reason that different assets have different betas is that they have different loadings.

- Smart-beta ETFs : instead of tracking a broad market index using market capitalization weights, they are funds designed to provide exposure to specific characteristics such as
    - size (SMB) and value (HML)
    - momentum (WML, for Winners Minus Losers), which is the return on a portfolio that buys recent well-performing stocks and sells poorly performing ones
    - volatility, as measured by the standard deviation of stock returns
    - quality, the difference in returns of stocks with high versus low return on assets or similar measures of profitability;
    - investment, the difference between returns on firms with high versus low rates of asset growth

## **A Multiperiod Fallacy**
the Markowitz model keeps giving the same weights, but the market portfolio weights change every period
1. If it is assumed that the statistical properties of the returns for the next period are identical to those of the previous period and the new returns are uncorrelated with those of the previous period, *the new weighting vector w will be equal to that of the previous period*
2. However, in the meantime the prices will have changed relative to each other; and hence the vector w will no longer correspond to the market portfolio since the market weights are capitalization weights, and a *price variation changes the capitalization*
3. In general, as prices change relative to each other, the dollar proportions represented in the market also change; but a repeating meanâ€”variance model dictates
that the dollar proportions of an optimal portfolio should remain fixed, which is a contradiction

Solutions
- assume that the expected returns change each period in a way that keeps the market portfolio optimal
- develop a full multiperiod approachthe multiperiod theory which suggests that price volatility is actually desirable, rather than undesirable
