## **Ch2 The Size Factor**

##### **Size Factor**
Calculated by taking the annual average return of small-cap stocks and subtracting the annual average return of large-cap stocks

e.g.) *SMB*(small minus big)
- Small-cap stocks are the stocks within deciles 6–10 of the CRSP index
- Large-cap stocks are those within deciles 1–5 of the CRSP index
- From 1927 through 2015, the U.S. size premium was 3.3 percent

##### **Sharpe ratio**
of the size premium  *0.24*
- Alternate definitions of the size premium can be notably higher
- Combining size with other factors can also be very effective

##### **Persistent**
the chances of success increase as the time period lengthens, although with somewhat lower values

##### **Intuitive**
Relative to large companies, small companies typically are characterized by:
- *greater leverage*
- *smaller capital base*, reducing their ability to deal with economic adversity;
- *greater vulnerability* to variations in credit conditions due to more restrictive access to capital;
- *higher volatility of earnings*;
- *lower levels of profitability*;
- *greater uncertainty of cash flow*; and *less liquidity*, which therefore makes their stocks more expensive to trade
- a less-proven, or even unproven, track record for the business model;
- less depth of management

Size premium is compnsation for its riskiness
- Small-cap stocks are more *volatile* than large-cap stocks
    - (1927 ~ 2015) annual standard deviation of small-cap stocks 30% **VS** 20% for large-cap stocks
- Smaller companies tend to perform relatively *poorly in bad times*
    - (1931, Great Depression) *CRSP 1–5 Index of large-cap stocks* lost 43.3% **VS** *CRSP 6–10 Index of small-cap stocks* lost 50.2 percent
    - (1973 ~ 1974, recession) large-cap stocks lost 39.2% **VS** small-cap stocks lost 53.1%
    - (2008) large-cap stocks lost 36.5% **VS** small-cap stocks lost 38.7%
- *Monetary policy* has a significant impact on the size effect (Gerald Jensen and Jeffrey Mercer, 2002, “*Monetary Policy and the Cross-Section of Expected Stock Returns*”)
    - When size is isolated, there is a significant small-firm premium only in periods of expansionary monetary policy
    - In restrictive periods, the size effect is not statistically significant
- Relationship between firm size and performance across the *economic cycle* (Moon K. Kim and David A. Burnie, 2002, “*The Firm Size Effect and the Economic Cycle*”)
    - Small companies grow faster than large companies in good economic times (their risk is rewarded)
    - do poorly in the worst of times (their risk materializes, frequently ending in bankruptcy)
    - The authors concluded that the size effect is really compensation for *economic cycle risk*
    - size premium should *vary across economic cycles*
- Returns of small-cap stocks are more *pro-cyclical* than the returns of large-cap stocks (Motohiro Yogo, 2006, “*A Consumption-Based Explanation of Expected Stock Returns.*”)

##### **Small-Cap Growth Anomaly**
The negative premium provided by “black hole” small-cap growth stocks

While small-cap stocks as a whole have provided higher returns (the size premium), **small-cap growth stocks have produced below-market returns**
- (1927 ~ 2015) U.S. small-cap growth stocks 8.7 % return**VS** 2 percent below the 11.8 percent return of U.S. smallcap stocks **VS** U.S. largecap stocks 9.8% return(Using Fama-French indices)
At the same time they earned lower annualized returns, small-cap growth stocks exhibited *higher volatility*
- (1927 ~ 2015) annualized standard deviation of returns 32 % **VS** 30 % for small-cap stocks as a whole

Behavioral Explanation

Nicholas Barberis and Ming Huang, 2008, “*Stocks as Lotteries: The Implications of Probability Weighting for Security Prices*”
- Investors have a preference for *positive skewness*
    - Such investments provide the small chance of a *huge payoff* (winning the lottery) which Investors find this small possibility *attractive*
    - Positively skewed securities tend to be *“overpriced”* and thus subsequently earn negative average excess returns
    - e.g.) low average return on initial public offerings(*IPOs*), private equity, and *distressed stocks*, despite their high risks

In theory, we would expect anomalies to be arbitraged away, but *in the real world, anomalies can persist* because there are *limits to arbitrage*
- many institutional investors are *prohibited* by their charters from taking short positions
- *the cost of borrowing* a stock in order to short it can be expensive
- *limited supply* of stocks available to borrow
- *unwilling to accept the risks of shorting* because of the potential for *unlimited losses*
- run the risk that their borrowed securities are *recalled* before the strategy pays off
- run the risk that strategy performs poorly in the short run, triggering an *early liquidation*

##### **Controling for Junk**
while size is powerful on its own, its merits further increase when *considered in combination with other factors*
- both the *value* and *momentum factors* are *stronger for small stocks* than for large stocks

Clifford Asness, Andrea Frazzini, Ronen Israel, Tobias Moskowitz, and Lasse Pedersen, 2015, “*Size Matters, If You Control Your Junk.*”
- Examined the size premium while *controlling for the quality factor*
- Stocks with very poor quality (i.e., “junk”) are typically very small, have low average returns, and are typically distressed and illiquid securities
    - These characteristics *drive the strong negative relation between size and quality*
    - the returns of these junk stocks chiefly explain the sporadic performance of the size premium
- *Small quality stocks outperform* large quality stocks and small junk stocks outperform large junk stocks
- Returns to small stocks would be notably higher except that they tend to be of lower quality
- The authors concluded that after controlling for quality, a significant size premium emerges :
    - stable through time(*persistent*);
    - robust to its specification(*robust*);
    - more consistent across seasons and markets(*pervasive*);
    - *not concentrated* in microcaps(*pervasive*);
    - *not captured by an illiquidity premium*;
    - explains interactions between size and other return characteristics, such as value and momentum;
    - results in an almost perfect monotonic relationship between size deciles and excess returns

Asness, Frazzini, Israel, Moskowitz, and Pedersen
- Examined the size premium while controlling for the *low-beta factor*
- High-beta stocks 
    - tend to be more speculative(lottery tickets) and have very poor historical returns
    - tend to be the very same stocks as those with low-quality attributes
- Small-cap stocks have negative exposure to two relatively new factors, the *profitability factor*(*RMW, robust minus weak*) and the *investment factor*(*CMA, conservative minus
aggressive*)
    - High-profitability firms outperform low-profitability ones
    - Low investment firms outperform high-investment ones
    - Controlling for these factors further improves the performance of size


## **Ch3 The Value Factor**

##### **Value Factor**
Eugene Fama and Kenneth French, 199, *“The Cross- Section of Expected Stock Returns”*
- Development of the *Fama–French three-factor model*
- Added the *size* and *value factors* to the market beta factor
    
Value Factor : Tendency for relatively cheap assets to outperform relatively expensive ones
- commonly used metrics for determining value academically is the *BtM ratio*
- *HML*
    - Return on high (the “H”) book-to-market (BtM) stocks minus the return on low (the “L”) BtM stocks
    - Value stocks are defined as the 30 percent of stocks with the highest BtM ratio 
    - Growth stocks are defined as the 30 percent of stocks with the lowest BtM ratio
- *Value premium* calculated by taking the annual average return of value stocks and subtracting the annual average return of growth stocks
    - (1927 ~ 2015) the annual U.S. value premium has been 4.8 percent
    
##### **Persistent**
Persistent of the value premium over the period (1927 ~ 2015)<Br/>
    
- Sharpe ratio of the value premium was 0.34

##### **Pervasive**
- (1975 ~ 2015) *the Fama-French International Growth Index* 8.6% return **VS** *Fama-French International Value Index* 13.8% return
- (1989 ~ 2015) *the Fama-French Emerging Markets Growth Index* 9.3% return **VS** *the Fama-French Emerging Markets Value Index* 13.0 % return
- (1982 ~ 2014) Value stocks in 15 European markets showed a sizeable return premium over growth stocks, value premium was 4.9 percent as a whole<br/>(DFA, 2015, *“Dimensions of Equity Returns in Europe”*)
- Significant return premium to value in 18 developed market countries(including US, UK, continental Europe, and Japan) with the strongest performance in Japan<br/>(Clifford Asness, Tobias Moskowitz, and Lasse Pedersen, 2013, *“Value and Momentum Everywhere"*)

##### **Intuitive**
Risk-based Explanation
- Value stocks are cheap because they tend to be firms in *distress, with high leverage, and that face substantial earnings risk*<BR/>(Nai-fu Chen and Feng Zhang, 1998, *“Risk and Return of Value Stocks.”*)
    - Examined three intuitive measures of distress present in value companies
        1. Cutting dividends by at least 25 percent
        2. a high ratio of debt to equity
        3. a high standard deviation of earnings
        - The three measures all captured the returns information (produced high correlations) contained in portfolios as ranked by BtM
        - When these three metrics were present, returns were greater
    - Investment is irreversible, once production capacity is put in place, it is very hard to reduce
        - Value companies carry more nonproductive capacity than growth companies
        - In periods of low economic activity, companies with nonproductive capacity(value companies) suffer a greater negative impact in earnings because the burden of nonproductive capacity increases, and they find it more difficult than growth companies to adjust capacity
        - In periods of high economic activity, the previously nonproductive assets of value companies become productive, while growth companies find it harder to increase capacity - In good times, capital stock is easily expanded. In bad times, adjusting the level of capital is an extremely difficult task — especially so for value companies

- Compensation for *higher vulnerability to financial distress*<BR/> (Motohiro Yogo, 2006, *“A Consumption-Based Explanation of Expected Stock Returns.”*)
    - Value stocks deliver *low returns during recessions*
    - Returns of value stocks are more *pro-cyclical* than growth stocks

- BtM effect is mostly a *leverage (risk) effect*<BR/>(Robert F. Peterkort and James F. Nielsen, 2005, *“Is the Book-to-Market Ratio a Measure of Risk?”*)
    - Positive relationship between higher stock returns and market leverage
    - When compared with market leverage, the BtM ratio *added* a small amount of explanatory power for stock returns
    - Considering only companies that they called “all-equity” firms (those with minimal amounts of mostly current debt, as opposed to long-term liabilities), there was *no BtM effect at all*

- Value companies tend to be firms *under distress, with high leverage and high uncertainty of cash flow*<BR/>(Ralitsa Petkova, 2006, *“Do the Fama-French Factors Proxy for Innovations in Predictive Variables?”*)
    - Shocks to the *default spread*(the spread between bonds of higher-rated credit and lower-rated credit) explain the cross-section of returns and is *consistent with value being a measure of distress risk*
    - Shocks to the term spread (the difference between shortterm bonds and long-term bonds) also explain the cross-section of returns and is further consistent with value being a measure of distress risk
        - Growth stocks are high-duration assets (much of their value comes from expected future growth), making them similar to long-term bonds
        - Value stocks are low-duration assets, making them more similar to short-term bonds
    Mohammed Elgammal and David G. McMillan, 2014, *“Value Premium and Default Risk,”*
    - (1927 ~ 2011) Positive association between the default premium and the value premium accompanied with evidence for a leverage effect on the value premium
    - Default premium captures systematic risk in the macroeconomy and that the value premium is associated with rational decision making on the part of investors
        - Value stocks characterized by poor performance, earnings and profitability compared with growth stocks are *more vulnerable to the risk of default* and lead the investors to require a higher return on value stocks as leverage increases
        - they are cheap for a reason — a reason related to their riskiness

- Value premium is largely based on *fundamental risk factors* within the economy and arises through macroeconomic risk<BR/>(Angela J. Black, Bin Mao, and David G. McMillan, *“The Value Premium and Economic Activity: Long-Run Evidence from the United States.”*)
    - value stocks are more sensitive to bad economic news, whereas growth stocks are more sensitive to good economic conditions
    - (1959 ~ 2005) Examined the relationship between the value premium and *macroeconomic variables* : *industrial production, inflation, money supply, and interest rates*
    1. Industrial Production
        - Industrial production rises(economic expansion)
            - Value stocks become less risky relative to growth stocks
            - The prices of value stocks increase more than the prices of growth stocks
            - Spread between the high book-to-market and low book-to-market stocks narrows and the value premium declines
        - Industrial production falls(economic contraction)
            - Value stocks become riskier relative to growth stocks
            - Prices of value stocks decrease faster than growth stocks, and the value premium increases (a sign of increased risk)
            - negative relationship between the value premium and industrial production
            - This certainly was the case in the most recent recession(2007 ~ 2009, the value premium was -0.44 percent per month)
    2. Money Supply
        - Negative relationship exists between the value premium and the money supply
        - increase in the money supply, stock prices increase, 
            - Prices of value stocks tend to increase more than growth stocks, and the value premium shrinks
        - When money supply decreases, stock prices decrease,
            - Prices of value stocks decreasing more than growth stocks, and the value premium increases
    3. Interest Rates
        - positive relationship between the value premium and interest rates
        - As long-run interest rates rise, stocks become less attractive than bonds and stock prices decrease
            - Prices of value stocks decreasing faster than the prices of growth stocks
            - Increasing value premium
        - When interest rates fall,
            - Prices of value stocks increase faster than the prices of growth stocks
            - Decreasing value premium

- Value stocks have been more volatile than the market
    - (1927 ~ 2015) Annual standard deviation of the* Fama-French Large Cap Index* was 19.7% **VS** *Fama-French Large Cap Value Index (exutilities)* 26.8%, *The Fama-French Large Cap Growth Index (ex-utilities)* 21.5%
    - Annual standard deviation of *the Fama-French Small Cap Index* was 30.1% **VS** 33.4% for both their Small Cap Value Index (ex-utilities) and Small Cap Growth Index(ex-utilities)

Behavioral Explanation(Mispricing)
- Investors are systematically too optimistic in their expectations for the performance of growth companies and too pessimistic in their expectations for value companies<BR/>(Josef Lakonishok, Andrei Shleifer, and Robert W. Vishny, 1994, *“Contrarian Investment, Extrapolation, and Risk”*)
- Investors confuse familiarity with safety
    - more familiar with popular growth stocks, those stocks tend to be overvalued
- Performance expectations for value firms may be too *pessimistic*<BR/>(Joseph D. Piotroski and Eric C. So, *“Identifying Expectation Errors in Value/Glamour Strategies: A Fundamental Analysis Approach,”*)
    - BtM ratio reflects the *market’s expectations about future performance*
    - Firms with low book-to-market ratios and low FSCOREs (weak fundamentals) were persistently overvalued, and firms with high book-to-market ratios and high FSCOREs (strong fundamentals) were persistently undervalued
        - (1972 ~ 2010) Value strategies would be *successful if prices do not accurately reflect the future cash flow* implications of historical information in a timely manner, resulting in equity prices that temporarily drift away from their fundamental values
        - Value/glamour effect was *concentrated* among the subset of firms where expectations implied by book-to-market ratios were *not aligned* with the strength of the firms’ fundamentals(e.g. FSCORE, which is based on nine financial signals designed to measure three different dimensions of a firm’s financial condition: profitability, change in financial leverage/liquidity, and change in operational efficiency)
        - Value/glamour effect was *nonexistent* among firms where expectations in price were *aligned* with the strength of the firm’s recent fundamentals
        - Returns to this “incongruent value/glamour strategy” are *robust and significantly larger* than the average return generated by a traditional value/glamour strategy
    - Performance expectations for value firms may be too *pessimistic* reflect improvements in fundamentals too slowly
        - Investors in *glamour stocks are likely to under-react to information that contradict their beliefs* about firms’ growth prospects
        - Value stocks, being inherently more distressed than glamour stocks, tend to be neglected by investors 
- *Anchoring* (Anderson and Zastawniak)
    - glamour investors anchor on the high P/E value of growth shares, while ignoring the high likelihood that this P/E ratio will change in the future
    - Examining the two extremes of stocks with positive earnings, value stocks outperformed glamour stocks by an annual average of 7.5 percent, but, their standard deviation of returns is only slightly greater than that of the glamour bin’s and not enough to account for the higher returns
    - Returns for glamour investors are relatively poor, whatever their time horizon. However, *value investors can expect superior returns if they hold for two to three years*(The returns to value stocks, while just 5 percent for the first year, were 21 percent for the second year, and 15 percent for the third year. After that, returns declined to little more than the returns on glamour shares.)
    - A glamour company has a 34 percent chance of becoming a loss-maker next year, compared to just 25 percent for a value company.
    - Extreme *loss-makers and value companies are most likely to remain* in their same bins the following year, at 32 percent and 34 percent, respectively. Companies in the bins between the extremes are much more likely to move. But glamour investors appear to *underestimate the tendency of their preferred stocks to change bins*.
    - Extreme loss-makers find it particularly difficult to break out of the spiral. They have only a one-in-six chance of turning a profit in the next year, compared to a 27 percent chance of being delisted. 
    - *Glamour stocks that remain in the same bin provide three times the rate of return of value stocks that remain in the same bin* (36 percent versus 12 percent, respectively). However, few companies stay in the same bin the following year, and *glamour stocks have a greater tendency to move* (roughly five-sixths of the time) and then give very poor returns. In addition, if glamour stocks start making losses, they suffer greatly. On the other hand, *value shares are much more likely to remain* in or near the value P/E range.
    - Smaller companies are much more likely than large companies to move from the glamour bins to the value bins. In other words, the *small high-flyers have a greater tendency to see their shares come tumbling down*, moving from lofty valuations to being more reasonably priced.
    - Shares in the extreme-loser bins have respectable mean returns, but the median returns are very poor and the standard deviation of returns is double what it is for the glamour and value bins. Any good returns from an extreme-loser stock depend on the firm becoming profitable, or at least halting such heavy losses. Each year, however, there is an almost 60 percent chance of an extreme-loser firm either ceasing to be quoted or remaining an extreme loser stock.
- *Loss Aversion* : Tendency to be more sensitive (place greater utility) on losses than on gains<br/>(Nicholas Barberis and Ming Huang, 2001, *“Mental Accounting, Loss Aversion, and Individual Stock Returns,”*)
    - Size of investor loss aversion depends on whether the individual has recently experienced gains or losses
    - Growth stocks are generally companies that have *performed well recently*, as evidenced by their current high price. Investors, therefore, become *less concerned* about future losses, being cushioned by the gains from recent performance. They thus apply a *lower risk premium* (as they are now willing to accept more risk) to growth stocks.Low required risk premium drives prices up even further future expected returns lower.
    - Value stocks are generally associated with companies that have *performed poorly in the recent* past, as evidenced by their current low prices. The *pain of the recent loss* causes investors to perceive these stocks as even riskier. They therefore *raise the required risk premium*, *driving prices even lower and expected future returns even higher*.

##### **Robust**
- While the most common value metric is the BtM ratio, there are other metrics that can be used to separate cheap from expensive stocks
    - (1952 ~ 2015, US) the annualized value premium as measured by BtM was 4.1 percent (t-stat3 = 2.4), 4.7 percent (t-stat = 2.9) as measured by the *cash flow-to-price ratio*, and 6.3 percent (t-stat = 3.4) as measured by the *earnings-to-price ratio*
- Value premium was there across all market capitalizations (although it has been greater in small-cap stocks than in large-cap stocks), in the non-U.S. developed markets, and in emerging markets as well (where the value premium was the largest)
    - (1980 ~ 2014, developed markets) Using the BtM metric, the premium was 6.1 percent, the earnings-to-price metric, 7.3 percent, cash flow-to-price metric, 8.0 percent<bR/>(Brandes Institute, *“Value vs. Glamour: A Long-Term Worldwide Perspective”*)
