## Ch1 Consumption-Based Model and Overvie

### **Consumption-Based Model**
An investor must decide how to allocate their resources between consumption and saving(investment)
- Saving is the future consumption benefiting from the returns along original savings
- The marginal utility of present consumption should be equal to the marginal utility of future consumption with the benefit of the additional returns generated by saving
- *The asset’s price should equal the expected discounted value of the asset’s payoff, using the investor’s marginal utility to discount the payoff*<br/>i.e.) If the price and payoff do not satisfy this relation, the investor should buy more or less of the asset

**Perspective**
- *Interest rates* are related to expected marginal utility growth, and hence to the expected path of consumption<br/>i.e.) In a time of high real interest rates, it makes sense to save, buy bonds, and then consume more tomorrow. Therefore, high real interest rates should be associated with an expectation of growing consumption
- Risk corrections to asset prices should be driven by the covariance of asset payoffs with marginal utility and hence by the *covariance of asset payoffs with consumption*<br/>i.e.) An asset that does badly in states of nature like a recession is less desirable than an asset that does badly in states of nature like a boom, former asset will sell for a lower price reflecting a discount for its 'riskiness' and this riskiness depends on a co-variance, not a variance
- Consumption is also low and marginal utility is high when the investor’s other assets have done poorly; thus we may expect that prices are low for assets that covary positively with a large index such as the market portfolio(*Capital Asset Pricing Model*)

### **Basic Pricing Equation**
$$p_t = E_t\left(\beta\frac{u'(c_{t+1})}{u'(c_t)}x_{t+1}\right)$$
where
- $c_t$ denotes consumption at period $t$
- $u$ is utility function<br/>
    - increasing, reflecting a *desire for more consumption*
    - concave, reflecting the *declining marginal value of additional consumption*
    - power utility form $$u(c_t) = \frac{1}{1-\gamma}c_t^{1-\gamma}$$ is often used. As $\gamma \to 1$ it become $u(c) = \ln(c)$
- $x_{t+1}$ is random variable of payoff in the next period of the investment made in period t<br/> i.e.) for stocks, payoff is its new price plus dividened, $x_{t+1}=p_{t+1}+d_{t+1}$ 

**Utility function over Periods** $$U(c_t, c_{t+1}) = u(c_t) + \beta E_t\left(u(c_{t+1})\right)$$ where $\beta$ is the subjective discount factor
- Curvature of the utility function generates *aversion to risk* and to *intertemporal substitution*<br/>i.e.) The investor prefers a consumption stream that is steady over time and across states of nature

**Proof to Basic Pricing Equation**

Denote by $\xi$ the amount of the asset he chooses to buy. Then, his problem is
\begin{align*}
\max_\xi \quad & u(c_t) + E_t\left(\beta u(c_{t+1})\right) \\
\text{s.t.} \quad & c_t = e_t - p_t \xi \\
& c_{t+1} = e_{t+1} + x_{t+1}\xi
\end{align*}
where $e_t$, $e_{t+1}$ are the income in period t, t+1.<br/>
Substituting the constraints into the objective, and setting the derivative with respect to $\xi$ equal to zero, we obtain the first-order condition for an optimal consumption and portfolio choice,
$$p_t u'(c_t) = E_t\left(\beta u'(c_{t+1})x_{t+1}\right),$$
or equivalently,
$$p_t = E_t\left(\beta \frac{u'(c_{t+1})}{u'(c_t)}x_{t+1}\right).$$
The investor buys more or less of the asset until this first-order condition holds.
- Standard marginal condition for an optimum : $p_tu'(c_t)$, represents the loss in utility an investor incurs if they buy another unit of the asset. $E_t [\beta u'(c_{t+1})x_{t+1}]$, represents the increase in (discounted, expected) utility they get from the extra payoff at $t+1$. The investor will continue to buy or sell the asset until the marginal loss equals the marginal gain
- Given the payoff $x_{t+1}$ and the investor's consumption choices $c_t$ and $c_{t+1}$, it tells us what market price $p_t$ to expect

**The stochastic discount factor**

The stochastic discount factor $m_{t+1}$ is defined as
$$m_{t+1} \equiv \beta \frac{u'(c_{t+1})}{u'(c_t)}$$
Using this, we can rewrite the basic pricing formula as
$$p_t = E_t(m_{t+1}x_{t+1})$$
- If there is no uncertainty, $$p_t = \frac{1}{R_f}x_{t+1}$$ where $R_f$ is the gross risk-free rate and $1/R_f$ is the discount factor
- Riskier assets have lower prices than equivalent risk-free assets, so they are often valued by using risk-adjusted discount factors $$p^i_t = \frac{1}{R_i}\mathbb{E}_t\left[x^i_{t+1}\right]$$ risky asset $i$ must be discounted by an asset-specific risk-adjusted discount factor $\frac{1}{R_i}$
- Incorporation of all risk corrections can be done by defining a single stochastic discount factor—*the same one for each asset*—and putting it inside the expectation
- Correlation between the random components of the common discount factor $m$ and the asset-specific payoff $x_i$ generates asset-specific risk corrections
- $m_{t+1}$ can be also regarded as marginal rate of substitution, or pricing kernel, state-price density
- Notice that $p = E(mx)$ would still be valid if we changed the utility function, but we would have a different function connecting $m$ to data<br/>All asset pricing models amount to alternative ways of connecting the stochastic discount factor to data

**Prices, Payoffs, and Notation**

Dividing the payoff $x_{t+1}$ by the price $p_t$ to obtain a gross return R as
$$R_{t+1} \equiv \frac{x_{t+1}}{p_t}$$
Thinking of the gross return R as a payoff with price 1, it should obey
$$1 = E(mR),$$
where $m$ is the stochastic discount factor
- Returns are often used in empirical work because they are stationary over time<br/>Stationary : the statistical sense; they do not have trends and can be meaningfully taken an average. It does not mean constant.
- Dividing by dividends and creating a payoff of the form $$x_{t+1} = \left(1+\frac{p_{t+1}}{d_{t+1}}\right)\frac{d_{t+1}}{d_t}$$ corresponding to a price $p_t/d_t$ is a way to look at prices but still to examine stationary variables
- Prices and returns can be real or nominal;<br/>p = E(mx) can refer to either case. The only difference is whether we use a real or nominal discount factor.<br/>We can create real prices and payoffs to write $${p_t\over \Pi_t}=\mathbb{E_t}\left[\beta\frac{u'(c_{t+1})}{u'(c_t)}\frac{x_{t+1}}{\Pi_{t+1}}\right]$$ where $\Pi$ denotes the price level (cpi). Obviously, this is the same as defining a nominal discount factor by $$p_t = \mathbb{E_t}\left[\beta\frac{u'(c_{t+1})}{u'(c_t)}\frac{\Pi_t}{\Pi_{t+1}}x_{t+1}\right]$$

**The risk-free rate**

The risk-free rate is known ahead of time, so $p = E(mx)$ becomes $1 = E(mR_f) = E(m)R_f$, thus, $$R_f = \frac{1}{E(m)}$$ 

**Economics behind real interest rates**
- Without uncertainty<br/>Using power utility $u(c) = c^{-\gamma}$, $$R_f = \frac{1}{\beta}\left(\frac{c_{t+1}}{c_t}\right)^\gamma$$
    - Real interest rates are high when people are impatient, when $\beta$ is low<br/>i.e.) If everyone wants to consume now, it takes a high interest rate to convince them to save
    - Real interest rates are high when consumption growth is high<br/>i.e.) In times of high interest rates, it pays investors to consume less now, invest more, and consume more in the future. Thus, high interest rates lower the level of consumption today, while raising its growth rate from today to tomorrow.
    - Real interest rates are more sensitive to consumption growth if the power parameter $\gamma$ is large<br/>i.e.) If utility is highly curved, the investor cares more about maintaining a consumption profile that is smooth over time, and is less willing to rearrange consumption over time in response to interest rate incentives. Thus it takes a larger interest rate change to induce him to a given consumption growth. 
- With some uncertainty<br/>Let that consumption growth is lognormally distributed and define the log risk-free rate $r^f_t$ and subjective discount rate $\delta$ by $$r^f_t = \ln R^f_t;~~\beta = e^{-\delta}$$ and $\Delta$ denotes the first difference operator, $$\Delta\ln c_{t+1} = \ln c_{t+1} - \ln c_{t}$$ Then, since $$R^f_t = \frac{1}{E_t\left[\beta\left(\frac{c_{t+1}}{c_t}\right)^{-\gamma}\right]}$$ using the fact that normal $z$ means $$E\left(e^z\right) = e^{E(z)+(1/2)\sigma^2(z)}$$ we have $$R^f_t = \left(e^{-\delta}e^{-\gamma E_t(\Delta \ln c_{t+1}) + (\gamma^2/2)\sigma^2_t(\Delta\ln c_{t+1})}\right)^{-1}$$ the real risk-free rate equation becomes $$r^f_t = \delta + \gamma E_t(\ln c_{t+1}) - \frac{\gamma^2}{2}\sigma^2_t(\ln c_{t+1})$$
    - The same results as we had with the deterministic case(without uncertainty)
    - New $\sigma^2$ term captures precautionary savings<br/>i.e. 1) When consumption is more volatile, people with this utility function are more worried about the low consumption states than they are pleased by the high consumption states. Therefore, people want to save more, driving down interest rates.<br/>i.e. 2) backwards: consumption growth is high when real interest rates are high, since people save more now and spend it in the future, and consumption is less sensitive to interest rates as the desire for a smooth consumption stream, captured by $\gamma$, rises.
    - the curvature parameter $\gamma$ simultaneously controls intertemporal substitution and risk aversion<br/>i.e. 1) Intertemporal substitution : aversion to a consumption stream that varies over time<br/>i.e. 2) Risk aversion : aversion to a consumption stream that varies across states of nature, and precautionary savings, which turns out to depend on the third derivative of the utility function

**Risk Corrections**

Using the definition of covariance $cov(m, x) = E(mx) - E(m)E(x)$, we can write $p = E(mx)$ as $$p = E(m)E(x) + cov(m, x)$$ Substituting the risk-free rate equation, we obtain $$p = E(x) \frac{1}{R_f} + cov(m, x)$$ 
- First term is the standard discounted present-value formula. This is the *asset’s price in a risk-neutral* world—if consumption is constant or if utility is linear.
- Second term is a *risk adjustment*. An asset whose payoff covaries positively with the discount factor has its price raised and vice versa
    - To understand the risk adjustment, substitute back for $m$ in terms of consumption, to obtain $$p = E(x) \frac{1}{R_f} + \frac{cov\left[\beta u'(c_{t+1}), x_{t+1}\right]}{u'(c_t)}$$
    - Marginal utility $u'(c)$ declines as $c$ rises. Thus, an asset’s price is lowered if its payoff covaries positively with consumption<br/>i.e.) Investors do not like uncertainty about consumption. If you buy an asset whose payoff covaries negatively with consumption, it helps to smooth consumption and so is more valuable than its expected payoff might indicate. ex) Insurance
- *Covariance of a payoff with the discount factor* rather than its variance determines its riskiness<br/>i.e. 1) The investor cares about the volatility of consumption. He does not care about the volatility of his individual assets or of his portfolio, if he can keep a steady consumption.<br/>i.e. 2) If the investor buys a little more $\xi$ of payoff $x$, $\sigma^2(c)$ becomes $$\sigma^2(c + \xi x) = \sigma^2(c) + 2\xi \text{cov}(c, x) + \xi^2\sigma^2(x)$$ For small (marginal) portfolio changes, the covariance between consumption and payoff determines the effect of adding a bit more of each payoff on the volatility of consumption.
- Restating the same intuition for the special case that the price is 1 and the payoff is a return, $$1 = E(mR^i)$$
which, asset pricing model is saying that, although expected returns can vary across time and assets, expected discounted returns should always be the same, 1.<br/>Applying the covariance decomposition $$1 = E(m)E(R^i) + cov(m, R^i)$$ and using $R_f = 1/E(m)$, $$E(R^i) − R_f = −R_f cov(m, R^i)$$ or $$E(R^i) − R_f = −{cov(u'(c_{t+1}), R^i_{t+1}) \over E(u'(c_{t+1}))}$$ All assets have an expected return equal to the risk-free rate, plus a risk adjustment<br/>i.e.) Assets whose returns covary positively with consumption make consumption more volatile, and so must promise higher expected returns to induce investors to hold them.


**Idiosyncratic Risk Does Not Affect Prices**

If the payoff $x$ is uncorrelated with the discount factor $m$, the asset will receive no risk correction to its price and will have an expected return equal to the risk-free rate. In mathematical terms, if $cov(m, x) = 0$, then $$p = \frac{\mathbb{E}(x)}{R_f}$$ regardless of the magnitude of $σ^2(x)$
- Idiosyncratic risk, uncorrelated with the discount factor, *generates no premium*
- *Only systematic risk*, a risk correctionthe component of a payoff perfectly correlated with the discount factor generates an extra return

The equation $x = proj(x|m) + \epsilon$ represents the projection of $x$ onto $m$, specifically, $proj(x|m) = \frac{\mathbb{E}(mx)}{\mathbb{E}(m^2)}m$.

By definition, the regression residuals are orthogonal to the right-hand side variable, which means that $\mathbb{E}(mε) = 0$.

This implies that the price of $\epsilon$ is zero, 

$$p(proj(x|m)) = \frac{\mathbb{E}(mx)}{\mathbb{E}(m^2)} = \frac{\mathbb{E}(m^2)\mathbb{E}(x)}{\mathbb{E}(m^2)} = \mathbb{E}(x)$$

The terms "systematic" and "idiosyncratic" are defined differently in different contexts
- In this context, *the residuals $\epsilon$ can be correlated with each other*, although they are not correlated with the discount factor
- *The Arbitrage Pricing Theory(APT)* begins with a factor-analytic decomposition of the covariance of payoffs, and the term "idiosyncratic" there is reserved for *the component of payoffs that is uncorrelated with all of the other payoffs*

**Beta Pricing Model : Expected Return-Beta Representation**

Rewriting $E(R^i) − R_f = −R_f cov(m, R^i)$ for a return $R^i$ as $$E(R^i) = R^f + \frac{cov(R^i,m)}{var(m)} \times \left(-\frac{var(m)}{E(m)}\right)$$ which simplifies to $$E(R^i) = R^f + \beta_{i,m} \times \lambda_m$$ where $\lambda_m = -\frac{var(m)}{E(m)}$ is often interpreted as the *price of risk* and the $\beta=\frac{cov(R^i,m)}{var(m)}$ as the *quantity of risk* in each asset.

Notice that the coefficient $\lambda_m$ is the same for all assets i, while the $\beta_{i,m}$ varies from asset to asset.

- Setting $m = \beta(c_{t+1}/c_t)^{-\gamma}$,<br/>we can take a Taylor approximation of equation to express betas in terms of consumption growth, rather than marginal utility. The result is $$E(R^i) = R^f + \beta_{i,\Delta c}\lambda_{\Delta c}$$ $$\lambda_{\Delta c} = \gamma\text{var}(\Delta c)$$
    - Expected returns should increase linearly with their betas on consumption growth itself
    - the factor risk premium $\lambda_{\Delta c}$ is determined by risk aversion and the volatility of consumption. The more risk averse people are, or the riskier their environment, the larger an expected return premium one must pay to get investors to hold risky (high beta) assets

**Mean-Variance Frontier**

All assets priced by the discount factor $m$ must obey $$E(R_i) - R_f \leq \frac{\sigma(m)}{E(m)}\sigma(R_i)$$ 

pf) writing for a given asset return $R_i$ $$1=E(mR_i)= E(m)E(R_i) + \rho_{m,R_i} \sigma(R_i)\sigma(m)$$
and hence $$E(R_i) = R_f - {\rho_{m,R_i}}\frac{\sigma(m)}{E(m)}{\sigma(R_i)}$$
- Means and variances of asset returns must lie in the wedge-shaped region, *mean-variance frontier*, illustrated in Figure
- The efficient frontier is generated by the condition $|\rho_{m,R_i}|=1$
- All frontier returns are also perfectly correlated with each other
- Given a mean-variance efficient return and the risk-free rate, we can find a discount factor that prices all assets and vice versa
- expected returns can be described in a single-beta representation using any mean-variance efficient return $$E(R^i) = R^f + \beta_{i,mv} \times (E(R^{mv}) - R^f)$$
- Consider a payoff $\frac{m}{E(m^2)}$. Its price is $\frac{E(m^2)}{E(m^2)} = 1$, so it is a return. It is on the mean-variance frontier. Thus, if we know $m$, we can construct a mean-variance efficient return.
- We can plot the decomposition of a return into a ‘‘priced’’ or ‘‘systematic’’ component and a ‘‘residual,’’ or ‘‘idiosyncratic’’ component as shown in Figure.

    

**Slope of the Mean-Standard Deviation Frontier**

The slope of the mean-standard deviation frontier is the largest available Sharpe ratio. It is given by $$\frac{E(R_{mv})-R_f}{\sigma(R_{mv})} = \frac{\sigma(m)}{E(m)} = \sigma(m)R_f$$ Therefore, the slope of the frontier is determined by the volatility of the discount factor.

Considering the power utility function $$u(c) = c^{-\gamma}$$ then, we have $$\frac{E(R_{mv})-R_f}{\sigma(R_{mv})} = \sigma\left(\frac{c_{t+1}}{c_t}\right)^{-\gamma}$$ If consumption growth is lognormal, we can approximate this as $$\frac{E(R_{mv})-R_f}{\sigma(R_{mv})} \approx \gamma\sigma\left(\ln\frac{c_{t+1}}{c_t}\right)$$
- Slope of the mean-standard deviation frontier is higher if the economy is riskier—if consumption is more volatile—or if investors are more risk averse. Both situations naturally make investors more reluctant to take on the extra risk of holding risky assets

**Equity Premium Puzzle**

In postwar U.S. data, *the slope of the historical mean-standard deviation frontier, or of average return-beta lines, is much higher than reasonable risk aversion and consumption volatility estimates suggest*. 
- Over the last 50 years in the United States, real stock returns have averaged 9% with a standard deviation of about 16%, while the real return on treasury bills has been about 1%. Thus, the historical annual market Sharpe ratio has been about 0.5
- Aggregate nondurable and services consumption growth had a mean and standard deviation of about 1%. We can only reconcile these facts if investors have a risk-aversion coefficient of 50!
- Recognizing market incompleteness makes matters worse. Aggregate consumption has about 0.2 correlation with the market return, while it has been assumed the worst possible case that consumption growth and asset returns are perfectly correlated. Adding this fact, you need risk aversion of 250 to explain the market Sharpe ratio!
- Individuals have riskier consumption streams than aggregate, but as their risk goes up their correlation with any aggregate must decrease proportionally, so to first order recognizing individual risk will not help either

1. people are a lot more risk averse than we might have thought
2. the stock returns of the last 50 years were largely good luck rather than an equilibrium compensation for risk
3. something is deeply wrong with the model, including the utility function and use of aggregate consumption data

**Random Walks and Time-Varying Expected Returns**

Going back to the basic first-order condition, $$ p_tu'(c_t) = E_t[\beta u'(c_{t+1})(p_{t+1}+d_{t+1})]$$ If investors are risk-neutral, i.e., $u(c)$ is linear or there is no variation in consumption, if the security pays no dividends between $t$ and $t+1$, and for short time horizons where $\beta$ is close to 1, this equations reduces to $$ p_t = E_t(p_{t+1})$$ Equivalently, prices follow a time-series process of the form $$ p_{t+1} = p_t + \epsilon_{t+1} $$ If the variance $\sigma^2_t(\epsilon_{t+1})$ is constant, prices follow a *random walk* or *martingale*. 
- If the price today is much lower than investors' expectations of the price tomorrow, then investors will try to buy the security, which will drive up the price of the security until the price today equals the expected price tomorrow
- Returns should not be predictable; dividing by $p_t$, expected returns $E_t(p_{t+1}/p_t) = 1$ should be constant; returns should be like coin flips
- The more generally the prices should follow a martingale after adjusting for dividends and scaling by marginal utility
- The random walk view has been remarkably successful
    - Since consumption and risk aversion do not change much day to day, we might expect the random walk view to hold pretty well on a day-to-day basis
- Contradicts the still popular notion that there are ‘‘systems’’ or ‘‘technical analysis’’
    - Despite decades of dredging the data, and the popularity of media reports that purport to explain where markets are going, trading rules that reliably survive transactions costs and do not implicitly expose the investor to risk have not yet been reliably demonstrated

More recently, evidence has accumulated that long-horizon excess returns are quite predictable.
- To some this evidence indicates that the whole enterprise of economic explanation of asset returns is flawed

To address this issue, writing our basic equation for expected returns as $$\begin{aligned}
{\operatorname{E}(R_{t+1})-R^f} &= -\frac{\operatorname{cov_t}(m_{t+1}, R_{t+1})}{\operatorname{E_t}(m_{t+1})} \
&= -\frac{\sigma_t(m_{t+1})}{\operatorname{E_t}(m_{t+1})} \sigma_t(R_{t+1})\rho_t(m_{t+1},R_{t+1}) \
&\approx \gamma_t \sigma_t(\Delta c_{t+1})\sigma_t(R_{t+1})\rho_t(m_{t+1},R_{t+1}),
\end{aligned} $$ where $\Delta c_{t+1}$ denotes consumption growth.

In the theory so far, including this equation, we have thought of an investor, today, forming expectations of payoffs, consumption, and other variables tomorrow. Thus, the moments are really all conditional. $E(x_{t+1})$ was used to denote conditional expectation; the notation $E(x_{t+1}|I_t)$ where $I_t$ is the information set at time $t$. In this equation, t subscripts to emphasize that the relation applies to conditional moments.

*Sometimes, the conditional mean or other moment of a random variable is different from its unconditional moment*
- In the special case that random variables are i.i.d., conditional and unconditional moments are the same
    - This is a special case and not likely to be true of asset prices, returns, and macroeconomic variables

We can see that returns can be somewhat predictable—the expected return can vary over time
- If the conditional variance of returns changes over time, we might expect the conditional mean return to vary as well
    - This explanation does not seem to help much in the data; variables that forecast means do not seem to forecast variances and vice versa
- Predictable excess returns have to be explained by changing risk—$\sigma_t(c_{t+1})$—or changing risk aversion $\gamma$
    - It is not plausible that risk or risk aversion change at daily frequencies, but fortunately returns are not predictable at daily frequencies
    - Risk and risk aversion change over the business cycle, and this is exactly the horizon at which we see predictable excess returns

**Present-Value Statement**
To relate a price to the entire cash flow stream, rather than just to one dividend and next period’s price, write out a longer-term objective, $$E_t \sum\limits_{j=0}^{\infty} \beta_ju(c_{t+j})$$ $$p_t = E_t \sum\limits_{j=1}^{\infty} \beta_j \frac{u'(c_{t+j})}{u'(c_t)} d_{t+j} = E_t \sum\limits_{j=1}^{\infty} m_{t,t+j} d_{t+j}$$ You can see that if this equation holds at time t and time t +1, then we can derive the two-period version $$p_t = E_t [m_{t+1}(p_{t+1} + d_{t+1})]$$ Thus, the infinite-period and two-period models are equivalent.

- ‘‘transversality condition’’ is required to dreive this equation $$\lim\limits_{j\to\infty} E_t[m_{t,t+j}p_{t+j}] = 0$$
    - This is an extra first-order condition of the infinite-period investor, which is not present with overlapping generations of two-period investors
    - Chaining the last equation together, you would get the previous equation of it with a extra term. The extra term can be ruled out by this condition
    - It rules out ‘‘bubbles’’ in which prices grow so fast that people will buy now just to resell at higher prices later, even if there are no dividends

we can write a risk adjustment to prices, as we did with one-period payoffs, $$p_t = \sum\limits_{j=1}^{\infty} \frac{E_t(d_{t+j})}{R^f_{t,t+j}} + \sum\limits_{j=1}^{\infty} cov_t(d_{t+j},m_{t,t+j})$$ where $R^f_{t,t+j} \equiv E_t(m_{t,t+j})^{-1}$ is the j-period interest rate. 

To write prices as a discounted value using a risk-adjusted discount factor $$p^i_t = \sum\limits_{j=1}^{\infty} E_t(d^i_{t+j}/(R_i)^j)$$

At a deeper level, the expectation in the two-period formula $p = E(mx)$ sums over states of nature. The longer-term objective equation is just sums over time as well and is mathematically identical.

**Discount Factors in Continuous Time**



The instantaneous total return is $$\frac{dp_t}{p_t} +\frac{D_t}{p_t}dt$$ We model the price of risky assets as diffusions, for example, $$\frac{dp_t}{p_t} = \mu(\cdot)dt + \sigma(\cdot) dz$$ where $\cdot$ denotes that the drift and diffusions $\mu$ and $\sigma$ can be functions of state variables, and $dz$ represents increments to a standard Brownian motion, with $z_{t+\Delta} - z_{t} \sim N(0, \Delta)$. 

The increments $\Delta$ are normal. However, the dependence of $\mu$ and $\sigma$ on state variables means that the finite-time distribution of prices $f(p_{t+\Delta}|I_{t})$ need not be normal.

Next, we need to express the first-order conditions in continuous time. The utility function is

$$U({c_t})=E_{t}\left[\int_{0}^{\infty}e^{-\delta t}u(c_{t})dt\right]$$

As we did in deriving the present-value price relation in discrete time, the first-order condition for this problem gives us the infinite-period version of the basic pricing equation right away,

$$\boldsymbol{p_tu'(c_t)=E_t\int_{s=0}^{\infty}e^{-\delta s}u'(c_{t+s})D_{t+s}ds \quad (1)}$$

This equation is an obvious continuous-time analogue to $p_t=E_{t}\sum_{j=1}^{\infty}\beta^{j}\frac{u'(c_{t+j})}{u'(c_t)}D_{t+j}$

It turns out that dividing by $u'(c_t)$ is not a good idea in continuous time, since the ratio $\frac{u'(c_{t+\Delta})}{u'(c_t)}$ is not well-behaved for small time intervals. Instead, we can keep track of the level of marginal utility. Therefore, define the "discount factor" in continuous time as

$$\boldsymbol{\Lambda_t=e^{-\delta t}u'(c_t) \quad (2)}$$

Then we can write the pricing equation as

$$p_t\Lambda_t=E_t\int_{s=0}^{\infty}\Lambda_{t+s}D_{t+s}ds$$

(Some people like to define $\Lambda_t=u'(c_t)$, in which case you keep the $e^{-\delta t}$ in the equation. Others like to scale $\Lambda_t$ by the risk-free rate, in which case you get an extra $e^{-\int_{\tau_0}^{s}r_{t+\tau}^fd\tau}$ in the equation. 

To derive this fundamental equation, take the difference of this equation at $t$ and $t+\Delta$(or, start directly with the first-order condition for buying the security at $t$ and selling it at $t+\Delta$),

$$p_t\Lambda_t=E_t\int_{s=0}^{\Delta}\Lambda_{t+s}D_{t+s}ds+E_t\left[\Lambda_{t+\Delta}p_{t+\Delta}\right]$$

For $\Delta$ small the term in the integral can be approximated as

$$p_t\Lambda_t \approx \Lambda_tD_t\Delta+E_t[\Lambda_{t+\Delta}p_{t+\Delta}]$$

$$0 \approx \Lambda_tD_t\Delta +E_t[(\Lambda_{t+\Delta}p_{t+\Delta}-\Lambda_tp_t)]$$

Taking the limit as $\delta \rightarrow 0$, 

$$\boldsymbol{0=\Lambda D{dt}+E_t\left[{d}(\Lambda p)\right] \quad (3)}$$

With no dividends and constant $ \Delta $, $ 0 = E_t(d_{p_t}) = E_t(p_{t+\Delta} - p_{t}) $ says that price should follow a martingale. Thus, $ E_t[d(\Lambda p)] = 0 $ means that marginal utility-weighted price should follow a martingale, and adjusts for dividends. Thus, it is the same as the equation $ p_t = E_t[m_{t+1}(p_{t+1} + d_{t+1})] $ that we derived in discrete time.

Breaking up the $ d(\Lambda_t p_t) $ term using Ito's lemma, $ d(\Lambda p) = p d{\Lambda} + \Lambda dp + dp d{\Lambda}$,
$$0 = \frac{D}{p}dt + E_t\left[\frac{d\Lambda}{\Lambda}+\frac{dp}{p}+\frac{d\Lambda}{\Lambda}\frac{dp}{p}\right]$$

Applying the basic pricing equations (1), (3) to a risk-free rate, we obtain

$$\boldsymbol{r_t^f dt = -E_t\left(\frac{d\Lambda_t}{\Lambda_t}\right) \quad (5)}$$

This equation is the obvious continuous-time equivalent to $R^f_t = \frac{1}{E_t(m_{t+1})}$

With this interpretation, we can rearrange equation (3) as

$$\boldsymbol{E_t\left(\frac{dp_t}{p_t}\right) + \frac{D_t}{p_t}dt = r_t^f dt - E_t\left[\frac{d\Lambda_t}{\Lambda_t}\frac{dp_t}{p_t}\right] \quad (4)}$$

This is the obvious continuous-time analogue to $E(R) = R_f - R_f \text{cov}(m,R)$

The last term in (4) is the covariance of the return with the discount factor or marginal utility. Since means are order $dt$, there is no difference between covariance and second moment in the last term of (4). The interest rate component of the last term of continuous-time analogue naturally vanishes as the time interval gets short.

With $\Lambda_t=e^{-\delta t}u'(c_t)$ we have

$$
\begin{aligned}
    d_t \Lambda_t&= -\delta e^{-\delta t}u'(c_t) dt + e^{-\delta t}u''(c_t)dc_t + \frac{1}{2}e^{-\delta t}u'''(c_t)dc_t^2 \\
    \frac{d\Lambda_t}{\Lambda_t} &= -\delta dt + \frac{c_tu''(c_t)}{u'(c_t)}\frac{dc_t}{c_t} + \frac{1}{2}\frac{c_t^2u'''(c_t)}{u'(c_t)}\frac{dc_t^2}{c_t^2}
\end{aligned}
$$

Denote the local curvature and third derivative of the utility function as

$$
\begin{aligned}
    \gamma_t &= -\frac{c_tu''(c_t)}{u'(c_t)} \\
    \eta_t &= \frac{c_t^2u'''(c_t)}{u'(c_t)}
\end{aligned}
$$

(For power utility, the former is the power coefficient $\gamma$ and the latter is $\eta_t=\gamma(\gamma+1)$.)

Using this formula we can quickly redo the relationship between interest rates and consumption growth,

$$
\begin{aligned}
    r^f_t &= -\frac{1}{dt} E_t \left( \frac{d\Lambda_t}{\Lambda_t} \right) \\
            &= \delta + \gamma_t \frac{1}{dt} E_t \left( \frac{dc_t}{c_t} \right) - \frac{1}{2}\eta_t\frac{1}{dt} E_t \left() \frac{dc_t^2}{c_t^2} \right)
\end{aligned}
$$

We can also easily express asset prices in terms of consumption risk rather than discount factor risk, 

$$
\begin{aligned}
    E_t \left( \frac{dp_t}{p_t} \right) + \frac{D_t}{p_t} dt - r^f_t dt &= \gamma E_t \left( \frac{dc_t}{c_t}\frac{dp_t}{p_t} \right)
\end{aligned}
$$

Assets whose returns covary more strongly with consumption get higher mean excess returns, and the constant relating covariance to mean return is the utility curvature coefficient $\gamma$

Since correlations are less than 1, this implies that Sharpe ratios are related to utility curvature and consumption volatility directly; 

Using $\mu_p \equiv E_t(d p_t/p_t)$, $\sigma_p^2 = E_t[(d p_t/p_t)^2]$, $\sigma_c^2 = E_t[(d c_t/c_t)^2]$,

$$
\begin{aligned}
    \frac{\mu_p + \frac{D_t}{p_t} dt - r^f_t dt}{\sigma_p} &\leq \gamma \sigma_c
\end{aligned}
$$

## Ch2 Applying the Basic Model

**Assumptions and Applicability**

Writing $p = E(mx)$, we do *not* assume
1. Markets are complete, or there is a representative investor
    - Complete markets/representative agent assumptions are used if one wants to use aggregate consumption data in $u'(c_t)$, or other specializations and simplifications of the model
2. Asset returns or payoffs are normally distributed (no options), or independent over time
3. Two-period investors, quadratic utility, or separable utility
    - Just interpret $u'(c_t)$ as the partial derivative of a general utility function with respect to consumption at time t, utility function doesn't have to be time- and state-separable utility
4. Investors have no human capital or labor income
    - By contrast, CAPM and ICAPM relies heavily on the assumption that the investor has no nonasset income
5. The market has reached equilibrium, or individuals have bought all the securities they want to
    - we can just interpret p = E(mx) as giving us the value, or willingness to pay for

i.e.) All of these assumptions may come later, in various special cases, but we have not made them yet. We do assume that the investor can consider a small marginal investment or disinvestment.

**General Equilibrium**

There is nothing that stops us from writing the basic pricing equation as $u'(c_t)=\mathrm{E_t}[\beta u'(c_{t+1})\frac{x_{t+1}}{p_t}]$
- We can also think of this equation as determining today’s consumption given asset prices and payoffs, rather than determining today’s asset price in terms of consumption and payoffs
- Thinking about the basic first-order condition in this way gives the permanent income model of consumption

Q. Asset returns and consumption: which is the chicken and which is the egg?

A. Neither, and for many purposes, it does not matter.
- The first-order conditions characterize any equilibrium; if you happen to know $E(mx)$, you can use them to determine $p$; if you happen to know $p$, you can use them to determine consumption and savings decisions

**Finding c and p in terms of truly exogenous forces**

1. Suppose that the production technologies are linear: the real, physical rate of return (the rate of intertemporal transformation) is not affected by how much is invested
    - consumption must adjust to these technologically given rates of return
    - specify the return process, and then solve the consumer’s portfolio and consumption rules<br/>e.g.) permanent income model, CAPM, ICAPM, the Cox, Ingersoll, and Ross(1985) model of the term structure

    

2. Suppost Nondurable consumption appears (or is produced by labor) every period. There is nothing anyone can do to save, store, invest, or otherwise transform consumption goods this period to consumption goods next period
    - ‘‘endowment economy’’
    - asset prices must adjust until people are just happy consuming the endowment process
    - consumption is exogenous and asset prices adjust<br/>Lucas (1978) and Mehra and Prescott (1985)

    

3. In the real economy, one can save or transform consumption from one date to the next, but at a decreasing rate : investment increases, rates of return decline
- the modeling with the linear technology (CAPM, CIR, permanent income) model, or the endowment economy model is still valid.
    - model economy as a linear technology, but choose for the rate of return on the linear technologies exactly the same stochastic process for returns that emerges from the general equilibrium. The resulting joint consumption-asset return process is exactly the same as in the original general equilibrium!
    - model endowment economy, but choose for the endowment process exactly the stochastic process for consumption that emerges from the equilibrium with a concave technology. The joint consumption-asset return process is exactly the same.
- Thus, there is nothing wrong in adopting one of the following strategies for empirical work:
    1. Form a statistical model of bond and stock returns, solve the optimal consumption-portfolio decision. Use the equilibrium consumption values in $p = E(mx)$
    2. Form a statistical model of the consumption process, calculate asset prices and returns directly from the basic pricing equation $p = E(mx)$
    3. Form a completely correct general equilibrium model, including the production technology, utility function, and specification of the market structure. Derive the  quilibrium consumption and asset price process, including $p = E(mx)$ as one of the equilibrium conditions
    - If the statistical models for consumption and/or asset returns are right, i.e., if they coincide with the equilibrium consumption or return process generated by the true economy, either of the first two approaches will give correct predictions for the joint consumption-asset return process
- The equilibrium consumption or asset return process one has modeled statistically may change in response to such changes in structure. For such questions one really needs to start thinking in general equilibrium terms.

**Consumption-Based Model in Practice**

The model  give a complete answer to all the questions of the theory of valuation with functional form for utility, numerical values for the parameters, and a statistical model for the conditional distribution of consumption and payoffs.

To be specific, consider the standard power utility function $u'(c) = c^{-\gamma}$.

Then, excess returns should obey $$0 = \mathbb{E_t} \left[\beta \left(\frac{c_{t+1}}{c_t}\right)^{-\gamma} R_{t+1}^e \right]$$

Taking unconditional expectations and applying the covariance decomposition, expected excess returns should follow $$E(R_{t+1}^e)=-R^f cov\left[\beta \left(\frac{c_{t+1}}{c_t}\right)^{-\gamma}, R_{t+1}^e \right]$$

Given a value for $\gamma$, and data on consumption and returns, you can easily estimate the mean and covariance on the right-hand side, and check whether actual expected returns are, in fact, in accordance with the formula.



- the model is not hopeless—there is some correlation between sample average returns and the consumptionbased model predictions
- But the model does not do very well. The pricing error (actual expected return − predicted expected return) for each portfolio is of the same order of magnitude as the spread in expected returns across the portfolios
- The consumption-based model is, in principle, a complete answer to all asset pricing questions, but *works poorly* in practice.

**Alternative Asset Pricing Models: Overview**
1. Different utility functions<br/> Which variables determine marginal utility is a far more important question than the functional form. Perhaps the stock of durable goods influences the marginal utility of nondurable goods; perhaps leisure or yesterday’s consumption affect today’s marginal utility. These possibilities are all instances of nonseparabilities. One can also try to use micro data on individual consumption of stockholders rather than aggregate consumption. Aggregation of heterogeneous investors can make variables such as the cross-sectional variance of income appear in aggregate marginal utility.

2. General equilibrium models<br/> General equilibrium models deliver equilibrium decision rules linking consumption to other variables, such as income, investment, etc. Substituting the decision rules $c_t = f(y_t, i_t, \ldots)$ in the consumption-based model, we can link asset prices to other, hopefully better-measured macroeconomic aggregates. In addition, true general equilibrium models completely describe the economy, including the stochastic process followed by all variables. They can answer questions such as why is the covariance ($\beta$) of an asset payoff $x$ with the discount factor $m$ the value that it is, rather than take this covariance as a primitive. They can in principle answer structural questions, such as how asset prices might be affected by different government policies or the introduction of new securities. 

3. Factor pricing models<br/>Model marginal utility in terms of other variables directly. Factor pricing models follow this approach. They just specify that the discount factor is a linear function of a set of proxies, $$m_{t+1} = a + b_A f_{A,t+1} + b_B f_{B,t+1} + \ldots$$ By and large, the factors are just selected as plausible proxies for marginal utility: events that describe whether typical investors are happy or unhappy. Many factor pricing models are derived as general equilibrium models with linear technologies and no labor income; thus they also fall into the general idea of using general equilibrium relations (from, admittedly, very stylized general equilibrium models) to substitute out for consumption.
    - the Capital Asset Pricing Model (CAPM) $$m_{t+1} = a + bR^W_{t+1}$$ where $R^W$ is the rate of return on a claim to total wealth, often proxied by a broad-based portfolio such as the value-weighted NYSE portfolio
    - The Arbitrage Pricing Theory (APT) <br/> use returns on broad-based portfolios derived from a factor analysis of the return covariance matrix
    - The Intertemporal Capital Asset Pricing Model (ICAPM)<br/> suggests macroeconomic variables such as GNP and inflation and variables that forecast macroeconomic variables or asset returns as factors
    - Term structure models such as the Cox-Ingersoll-Ross model<br/>specify that the discount factor is a function of a few term structure variables, for example the short rate of interest and a few interest rate spreads

4. Arbitrage or near-arbitrage pricing<br/> The mere existence of a representation $p = E(mx)$ and the fact that marginal utility is positive $m \geq 0$ can often be used to deduce prices of one payoff in terms of the prices of other payoffs
    - The Black-Scholes option pricing model <br/> Since the option payoff can be replicated by a portfolio of stock and bond, any discount factor m that prices the stock and bond gives the price for the option
