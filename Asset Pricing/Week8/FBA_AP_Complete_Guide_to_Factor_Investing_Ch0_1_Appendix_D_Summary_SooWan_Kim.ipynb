### **Ch1 Market Beta**

Market beta 
- expresses the degree to which an asset tends to move with the broad market
- defined mathematically as the correlation between the asset’s return and the market’s return, multiplied by the ratio of the asset’s volatility to the market’s volatility
- *persistent*
- *pervasive*

    
    
- *investable*
- *Intuitive*
    1. risk of owning equities is highly correlated with the risks of the economic cycle
    2. a large percentage of equities, if not the vast majority, are owned by high-net-worth individuals : as net worth increases, the marginal utility of wealth decreases, and when individuals attain a level of wealth at which there is no longer a need to take risk, only a large risk premium might be enough to induce them to take the risk
    3. investment life cycle and borrowing constraints : young investors has lower level of income, and older investors are also in the withdrawal stage of their investment life cycle. risk of holding equities is concentrated in the hands of the middle-aged, saving consumer, minimizing the risks of the aforementioned double whammy at a time when they are closer to retirement
    4. the volatility of stocks is much greater than it is for riskless one-month U.S. Treasury bills

CAPM flawed
- the CAPM explained only about *two-thirds of the differences* in returns of diversified portfolios
- anomalies that violated the CAPM began to surface. And eventually, other factors were “discovered.”

FAMA–FRENCH THREE-FACTOR MODEL
- better model could be built using more than just the single factor
- along with the market beta factor, exposure to the factors of size and value further explains the differences in returns of diversified portfolios
- accounting for more than 90 percent of the differences in returns between diversified portfolios

### **Appendix D. The Low-Volatility Factor**

Anomaly : Low Volatility Facor
- superior performance of low-volatility stocks
- *pervasive*
    - not only for stocks, but for bonds as well
    - low-volatility strategies extends to international equities as well
- *Intuitive*
    - empirical failure of the CAPM
        - constraints on either leverage or short-selling or aversion to its use
        - market frictions
        - assumptions that investors are risk-averse, maximize the expected utility of absolute wealth, and care only about the mean and variance of returns do not holds    
        - assumptions that investors have complete information, and that they rationally process the information available don't holds
    - investors looking to increase their return choose to tilt their portfolios *toward* high-beta securities to garner more of the equity risk premium
    - Regulators typically do not distinguish between different stock types, but merely consider the total amount invested in stocks for determining required solvency buffers -> Investors who wish to maximize equity exposure but minimize the associated capital charge under such regulations are drawn to the high-volatility segment of the equity market, as it effectively gives most equity exposure per unit of capital charge
    - *winner’s curse* : In a market with little or no short selling, the demand for a particular security comes from the minority who hold the most optimistic expectations about it. As divergence of opinion is likely to increase with risk, high-risk stocks are more likely to be overpriced than are low-risk stocks
    - investors subject to all sorts of biases
        - Agent's compensation arrangement resembles a call option on the portfolio return, the value of which can be increased by creating a more volatile portfolio
        - investors with a preference for lottery-like investments
            - pay a premium to gamble e.g.)IPOs
            - exhibit positive skewness and excess kurtosis 
        - prefer higher relative wealth over lower absolute wealth
        - tend to buy attention-grabbing stocks which are typically high-volatility stocks, while boring low-volatility stocks suffer from investor neglect
    - overconfident

*Understanding Defensive Equity*, Robert Novy-Marx(2016)
- stocks in highest-quintile of volatility dramatically underperform, while the performance of the other four quintiles are very similar and market-like
- high-volatility and high-market-beta stocks tilt strongly to small-cap, unprofitable, and growth firms
- a stock’s profitability is a significant negative predictor of its volatility, and it is the single most significant predictor of low volatility
- including profitability as a factor, *the defensive strategy*(long low-volatility/short high-volatility portfolios) performance is well explained by controlling for the common factors of size, profitability, and relative valuations
- defensive strategies tilt strongly toward large-cap stocks (SMB loading of -1.12), value stocks (HML loading of 0.42), and profitable stocks 
- value and profitability tend to be strongly negatively correlated -> unless you control for profitability, the value loadings of defensive strategies will be lowered
- defensive strategy *constructed within small-cap growth stocks* performence was best -> the success of defensive strategies in small-cap stocks is a result of
avoiding these high-risk stocks(poor returns to smallcap growth stocks drivend by unprofitable stocks that tend to have negative book equity)
- majority of the returns from low volatility (and low beta) come from the short side, not because of the low-volatility strategy itself

problems with implementing a low-volatility strategy
- the excess return defensive portfolio present only in the first month after formation, and that they are largely subsumed by *high transaction costs* associated with shorting low-liquidity stocks(such as lowpriced/ high-volatility stocks)
    - can use long-only portfolio with screening to eliminate stocks with low liquidity, or can use patient trading strategies to minimize turnover costs
- the average price of stocks in the highest-volatility quintile was just more than $7, indicating that many would be considered *“penny stocks”*
- the low-risk effect has been noticeably weaker since 1990
    - many high-beta stocks simply disappeared after the dot-com crash, and the number of stocks on public U.S. exchanges has decreased dramatically since then
- that both high-volatility stocks and stocks *with high short interest* exhibit poor risk-adjusted future performance
    - with *low short interest actually experience extraordinary positive returns*
    - with high short interest experience equally extraordinary negative returns

Term risk
- low-volatility strategies have exposure to term risk (the duration factor)
    - low-volatility/low-beta stocks has the characteristics of safety as opposed to risk and opportunity -> high correlations with long-term bond returns
- their positive exposure to term risk, low-volatility stocks have benefited from the cyclical bull market in bonds that we have experienced since 1982. We cannot be sure it will be repeated

*Low-Beta Strategy*
- *Betting Against Beta (BAB) factor* : a portfolio that holds low-beta assets, *leveraged to a beta of 1*, and that shorts high-beta assets, deleveraged to a beta of 1
- the BAB factor has highly significant risk-adjusted returns after accounting for its realized exposure to the market beta, value, size, momentum, and liquidity factors
- higher beta means high volatility, thus, low-volatility anomaly may support this strategy
- alphas and Sharpe ratios declined almost monotonically as beta increased for stocks, U.S. Treasury bonds, credit markets, and futures markets for currencies and commodities
- relative flatness of the security market line is a *pervasive* global phenomenon
- returns increased with volatility until the highest-volatility quintile, whose returns lagged the rest(consistent with reulst in low-volatility factor)

Potential Overgrazing
- Cash inflows have raised the valuations of defensive (low-volatility/low-beta) stocks, dramatically reducing their exposure to the value premium to near zero or negative, and thus lowering expected returns
- because there is an ex-ante value premium, what low volatility is predicting at this point in time is not higher returns, just low future volatility
- how big of an impact valuations will have on low-volatility strategies is ambiguous
    - low-volatility strategies exposure to the value factor is time-varying : low-volatility factor spends about 62 percent of times in a value regime and 38 percent of times in a growth regime and this regime-shifting behavior impacts the performance of low-volatility strategies
- However, when low volatility *has negative* exposure to the value factor (as it did in mid-2016), it also forecasts below-market returns

Conclusion
- *invest directly in size, value, and profitability*, rather than doing so indirectly (like defensive strategies do)
- the anomaly is much more about the underperformance of high-volatility (or high-beta) stocks and not so much about the outperformance of low-volatility (or low-beta) stocks. Thus, to take advantage of the anomaly, you can either *avoid high-volatility/high-beta stocks* or invest in funds that screen them out.
- With interest rates at historic lows, the term premium aspect is *unlikely* to pay off as it has in the past
- recent strong demand for low volatility has caused it to lose the value exposure typical of it over the last several decades
