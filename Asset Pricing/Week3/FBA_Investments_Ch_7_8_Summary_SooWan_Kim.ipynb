### Ch 7 Efficient Diversification

- ***Sources of Risk to Portfolio***
    - **Market Risk**
        - *nondiversifiable*
        - *systematic risk*
        - comes from *macroeconomic factors* conditions in the general economy, such as *business cycle, inflation, interest rates, and exchange rates*.
    - **firm-specific risk**
        - *diversifiable*
        - *unique risk*
        - comes from success *firm-specific influences*, such as success in *research and development and personnel changes*.

- ***Diversification***<br/> : holding **fixed the total funds** put at risk and then spreading exposure across multiple sources of uncertainty
    - General Formulas for Portfolio composed of n Risky Assets
        - Expected Return $$E(r_p)=\sum_{i=1}^nw_iE(r_i)$$ 
        - Variance $$\sigma_p^2 = \sum_{i=1}^n\sum_{j=1}^n w_iw_jCov(r_i,r_j)$$
    - Variance for Equally Weighted Portfolio ($w_i = {1 \over n}$)
        - Variance $$\sigma _p^2 = {1 \over n}\sum_{i=1}^{n} \sigma_i^2 + \sum_{j=1\\{j\neq i}}^{n} \sum_{{i=1}}^{n}{1 \over {n^2}}Cov(r_i,r_j)$$ If we define the average variance and average covariance of the securities as $${\bar\sigma}^2 = {1 \over n}\sum_{i=1}^n\sigma_i^2$$ $$\overline{Cov}={1 \over {n(n-1)}}\sum_{j=1\\j\neq i}^n\sum_{i=1}^nCov(r_i,r_j)$$ then, $$\sigma_p^2={1\over n}{\bar\sigma}^2+{{n-1}\over{n}}\overline{Cov}$$
        - The first term ${1 \over n}\bar\sigma^2$ represents *firm-specific risks* and approaches *zero* as diversification increases(n gets ever larger)
        - The second term ${{n-1}\over{n}}\overline{Cov}$ represents *markter risks* and approaches $\overline{Cov}$ as diversification increases(n gets ever larger)
        - Thus, portfolio standard deviation falls as the number of securities increases, but it cannot be reduced to zero.
        - The risk of a highly diversified portfolio depends on *market risk*, the covariance of the returns of the component securities which in turn is a *function of the importance of systematic factors* in the economy, while *firm-specific risks* is eliminated by diversification.
        

- ***Portfolios of Two Risky Assets***<br/>
    : $w_D$ invested in the bond fund($r_D, \sigma_D$), and $1  w_D=w_E$, is invested in the stock fund($r_E, \sigma_E$).<br/>
    - Rate of Return on the Portfolio $$r_p = w_Dr_D+w_Er_E$$
    - Expected Return on the Portfolio $$E(r_p) = w_DE(r_D)+w_EE(r_E)$$
    - Variance of Rate of Return ot the Portfolio $$\sigma_p^2=w_D^2\sigma_D^2 + w_E^2\sigma_E^2+2w_Dw_ECov(r_D,r_E)$$ Since, $$Cov(r_D,r_E)=\rho_{DE}\sigma_D\sigma_E$$, therefore,$$\sigma_p^2=w_D^2\sigma_D^2 + w_E^2\sigma_E^2+2\rho_{DE}\sigma_D\sigma_E$$   
        - portfolios of less than perfectly correlated assets(correlation coefficient of 1) has standard deviation less than that weighted average
        - portfolios of less than perfectly correlated assets *always* offer some degree of diversification benefit(the portfolios expected return is the weighted average of its component expected returns, whereas its standard deviation is less than the weighted average of the component standard deviations).
        - The lower the correlation, the greater the potential benefit from diversification.
        - hedge asset has negative correlation with the other assets in the portfolio.
    - Minimum-Variance Portfolio $$w_{Min}(D)=\frac{\sigma_E^2-Cov(r_D,r_E)}{\sigma_D^2+\sigma_E^2-2Cov(r_D,r_E)}$$
        - has a standard deviation *smaller than that of either of the individual component assets*.
    - Portfolio Opportunity Set
    
    - Asset Allocation : Construction of a risky portfolio combining the major asset classes that provides the highest possible Sharpe ratio.
        - the higher the Sharpe ratio, the greater the expected return corresponding to any level of volatility.
        - best *optimal risky portfolio* P is the one that results in the steepest capital allocation (CAL) line.
        - In the case of two risky assets, the weights of the optimal risky portfolio, P, are $$w_D=\frac{E(R_D)\sigma_E^2-E(R_E)Cov(R_D,R_E)}{E(R_D)\sigma_E^2+E(R_E)\sigma_D^2-[E(R_D)+E(R_E)]Cov(R_D,R_E)}\\ w_E=1-w_D$$
    
    - We can use constructed optimal risky portfolio, P, and its expected return and volatility along with the individual investors degree of risk aversion, A, to calculate the optimal proportion of the complete portfolio to invest in this risky component.
    

- ***The Markowitz Portfolio Optimization Model***<br/>: generalize the portfolio construction problem to the case of many risky securities and a risk-free asset<br/>
    0. Assume that analysts already have spent the time and resources to prepare the inputs
        - Input List : n estimates of $E(r_i)$ and the n  n estimates of the covariance matrix
            - n Rate of Returns, n SDs, $n(n-1) \over 2$ Covariances has to be calculated
            - The security valuation techniques and methods of financial analysis that analysts use *to be examined in later Chapter.*
        - **If sample average returns over recent years are used as proxies for the future return on the security, the noise in those estimates will make the resultant efficient frontier virtually useless for portfolio construction**
            - Consider a stock with an annual standard deviation of 30%. Even if one were to use a 10-year average to estimate its expected return (and 10 years is almost ancient history in the life of a corporation), the standard deviation of that estimate would still be $\frac{300}{\sqrt{10}} = 9.5%$. Given this level of imprecision, the chances that the sample average represents expected returns for the coming year are negligible
            - **Moreover, you cannot avoid this problem by observing the rate of return on the stock more frequently. The accuracy of the sample average as an estimate of expected return depends on the length of the sample period; it is not improved by sampling more frequently within a given sample period.**
        - May be subject to additional constraints
            - many institutions are prohibited from taking short positions(Short-Sale Restrictions) $=>$ add to the optimization program constraints that rule out negative (short) positions.
            - demand a minimal dividend yield $=>$ input list will be expanded to include the dividend yield, and the optimization program will include an additional constraint requiring the dividend yield of the portfolio to equal or exceed the desired level
            - SRI(socially responsible investing) : tend to impose more absolute restrictions on portfolio choice to reflect their ethical concerns
            - ESG(environmental, social, and governance-focused) : tend to be more focused on long-term sustainable business prospects that might be enhanced by environmentally and socially sound practices.
    1. Determine the riskreturn opportunities available
        - Get **Minimum-Variance Frontier** of risky assets
             - Minimum-Variance Frontier : graph of the lowest possible variance that can be attained for a given portfolio expected return
            - Given the input data for expected returns, variances, and covariances, we can calculate the minimum-variance portfolio consistent with any targeted expected return.
            - All the individual assets lie to the right inside the minimum-variance frontier, at least when we allow short sales in the construction of risky portfolios. Diversification allows us to construct portfolios with higher expected returns and lower standard deviations.
        - Get **Efficient Frontier** of risky assets
            - Efficient Frontier : portion of the frontier that lies above the global minimum-variance portfolio
            - For any portfolio on the lower portion of the minimum-variance frontier, there is a portfolio, with the same standard deviation and a greater expected return positioned directly above it. Hence *inefficient*.
        - Methodology
            1. Look for the portfolio that plots farthest to the left on horizontal line.
                1. draw the constraints, that is, horizontal lines.
                2. look for the portfolio with the lowest standard deviation that plots on each horizontal line(at the level of required expected returns)
                3. repeat this for many levels of required expected returns and connect the dots, the shape of the minimum-variance frontier emerges
                4. We then discard the bottom (dashed) half of the frontier because it is inefficient.
            3. Look for the portfolio that plots highest on this vertical line
                1. draw a vertical line that represents the standard deviation constraint. 
                2. consider all portfolios that plot on this line (have the same standard deviation) and choose the one with the highest expected return.
                3. Repeating this procedure for many vertical lines (levels of standard deviation) gives us the efficient frontier.
    
    2. Search for the capital allocation line with the highest Sharpe ratio
        - CAL generated by the optimal portfolio, P, is the one tangent to the efficient frontier.
        - Methodology
            1. ratchet up the CAL by selecting different portfolios until we reach portfolio P
            2. ask the spreadsheet program to maximize the Sharpe ratio of portfolio P(*can skip the charting of the efficient frontie*)
    
    3. The individual investor chooses the appropriate mix between the optimal risky portfolio P and risk-free asset(Capital Allocation).

- ***Optimal Portfolios and Non-Normal Returns***
    - Markowitz Portfolio Optimization Model assumes normal distributions of returns in that standard deviation is taken to be a fully adequate measure of risk.
    - potential non-normality of returns requires us to pay attention to risk measure, especially *worst-case losses such as value at risk(VaR) or expected shortfall(ES)*
        - forecasts of greater-than-normal VaR and ES should encourage more moderate capital allocations to the risky portfolio
        - accounting for the effect of diversification on VaR and ES would be useful though it cannot be easily estimated
        - a practical way to estimate values of VaR and ES in the presence of fat tails is *bootstrapping*<br/> : with a historical sample of returns of each asset in our prospective portfolio, compute the portfolio return corresponding to a draw of one return from each assets history, repeat these random draws, calculate hypothetical (but still empirically based) returns on as many of these random portfolio returns as we wish. Fifty thousand portfolio returns produced in this way can provide a good estimate of VaR and ES values.

- ***Risk Pooling, Risk Sharing, and Time Diversification***
    - if an insurance company sells $n$ identical fire insurance policies that the risk of a fire across different policy holders is uncorrelated($Cov=0$), and $n$ investors each owns $1\over n$ of the firm,
        - Variance of *total* payouts$$Var(\sum_{i=1}^nx_i) = n\sigma^2$$  **Risk Pooling** : pooling together many sources of independent risk sources so that total payouts variance increases with the number of policies, actually becoming more uncertain 
        - Variance of the *average* payoff to each investors  $$Var({1\over n}\sum_{i=1}^nx_i)={1\over{n^2}}\times n\sigma^2={{\sigma^2}\over n}$$ **Risk Sharing** : as $n$ increases, the variance of the *average* insurance policy payoff of each investor's exposure decreases approaching zero, while each investor's investment is unchanged.
        - Risk sharing is the essential complement to risk pooling preventing any individual's total risk from growing with the number of policies as each investor's exposure to any single policy shrinks
    - When market returns are independent from year to year, with a mean annual return of 5% and standard deviation 30% and assuming the risk-free rate is zero,<br/>
        - Expected average return is fixed at 5%, average standard deviation falls with the investment horizon
        - The probability that the average return will be positive increases as the time horizon extends(calculated assuming normal-distribution)
        - The standard deviation of cumulative returns increases, and the magnitude of underperformance is progressively worse(line 6 ~ 9)
        - **By extending horizon of investments, average return per year may now be more predictable, but the uncertainty of your cumulative return increases**
        - investment returns are not safer at long horizons, and **time diversification is only an illusion**.
        - **True diversification means holding <u>fixed the total funds put at risk</u>, and then spreading exposure across multiple sources of uncertainty.** Thus, true time diversification is spreading total fixed funds across investing periods(Investing \\$10000 in period 1 VS Investing \\$5000 in each period 1, 2).

### Ch 8 Index Models

##### Single-Factor Security Market

- Focus on separating risk by the actual rate of return on security, into the sum of its previously expected value plus an unanticipated surprise$$r_i=E(r_i)+~unanticipated~surprise$$
    - *Mean of unanticipated surprise = 0* (over time, surprises will average out to zero)
- **Single-Factor Model**$$r_i =E(r_i)+\beta_i m +e_i$$
    - Assumption
        - $E(r_i)$ : Originally Expected Return. fixed value
        - $m$ : Unanticipated Market Surprise(general macroeconomic surprise). systematic market factor random variable
            - mean : 0 
            - standard deviation : $\sigma_m$
        - $e_i$ : Unanticipated Firm-Specific Surprise. firm-specific random variable
            - mean : 0
            - standard deviation : $\sigma (e_i)$
        - $\beta_i$ : Sensitivity Coefficient for Firm $i$ to Macro Condition
            - firms have greater sensitivity to the general market economic conditions has higher betas <br/>ex) auto firm, cyclical firm
        - *$m$ and $e_i$ are assumed to be uncorrelated*<br/>($e_i$ which is firm-specific is independent of shocks to the common factor that affect the entire economy)
    - Total Variance $$\sigma_i^2 \beta_i^2\sigma_m^2+\sigma^2(e_i)$$
    - Covariance between Two Firms' Returns $$Cov(r_i,r_j)=Cov(\beta_im+e_i,~\beta_jm+e_j)=\beta_i\beta_j\sigma_m^2$$

##### The Single-Index Model

- Relation between the return on Sequrity and the Return on the Market Index
    - With a historical sample of paired observations of excess returns on the market index and a particular security, if we fit a straight line through scatter diagram :
    
        - Slope of the line reflects the sensitivity of Amazons return to market conditions : Steeper the line, the rate of return more responsive to the market return
        - The scatter of points around the line implies firm-specific events also have a significant impact on Amazons return (if returns perfectly tracked those of the market, then all return pairs would lie exactly on the line)
- **Single-Index Model** <br/>
    - For any stock $i$, denoting the pair of excess returns in month $t$ by $R_i(t)$ and $R_M(t)$, 
$$R_i(T)=\alpha_i+\beta_iR_M(t)+e_i(t)$$<br/>where,<br/>
        - $\alpha$(Vertical Intercept of Regressuin Equation) : security's expected excess return when the market excess return is zero
        - $\beta_i$(Slope of the Regression Equation) : measures the security's sensitivity to the market index
        - $e_i$(Residual of Regression) : zero-mean, firm-specific surprise in the security return in month t<br/>
    - Uses the market index to stand in for the systematic factor of single-factor model
    - Assume average beta of all stocks in the economy is 1
- Expected Return and Covariance in the Single-Index Model
    - Expected Return in the Single-Index Model<br/>Since $E(e_i)=0$, <br/>$$E(R_i)=\alpha_i+\beta_iE(R_M)$$
        - Systematic Risk Premium($\beta_iE(R_M)$) : part of a security's risk premium due to the risk premium of the market index
            - securities with high betas have a magnified sensitivity to market risk and will therefore enjoy a greater risk premium as compensation for this risk
        - Nonmarket Risk Premium($\alpha_i$) : remainder of the risk premium after removing systematic risk premium
            - $\alpha$ may be large if a security is underpriced and therefore offers an attractive expected return
    - Covariance in the Single-Index Model<br/>
Total Risk = Systematic risk + Firm-specific risk
$$\sigma_i^2=\beta_i^2\sigma_M^2+\sigma^2(e_i)$$
Covariance = Product of betas  Market-index risk
$$Cov(r_i,~r_j)=\beta_i\beta_j\sigma_M^2$$
Correlation = Product of correlations with the market index
$$Corr(r_i,~r_j)={\beta_i\beta_j\sigma_M^2 \over \sigma_i\sigma_j}={\beta_i\sigma_M^2\over\sigma_i\sigma_M}\times{\beta_j\sigma_M^2\over\sigma_j\sigma_M}=Corr(r_i,~r_M)\times Corr(r_j,~r_M)$$
- Diversification  
    - Excess Return on the Portfolio of Stocks$$R_P=\alpha_P+\beta_PR_M+e_P$$
    - Excess Rate of Return on the Equally Weighted Portfolio(each portfolio weight $w_i = {1\over n}$)
        $$R_P=\sum_{i=1}^nw_iR_i={1\over n}\sum_{i=1}^nR_i={1\over n}\sum_{i=1}^n(\alpha_i+\beta_iR_M+e_i)={1\over n}\sum_{i=1}^n\alpha_i+({1\over n}\sum_{i=1}^n\beta_i)R_M+{1\over n}\sum_{i=1}^ne_i$$, therefore,
        $$\beta_p={1\over n}\sum_{i=1}^n\beta_i$$
        $$\alpha_P={1\over n}\sum_{i=1}^n\alpha_i$$
        $$e_P={1\over n}\sum_{i=1}^ne_i$$
    - Variance of the Equally Weighted Portfolio(each portfolio weight $w_i = {1\over n}$)$$\sigma_P^2=\beta_P^2\sigma_M^2+\sigma^2(e_P)$$
        - systematic component of the portfolio variance, $\beta_P^2\sigma_M^2$, depends on portfolio beta and will persist regardless of the extent of portfolio diversification, therefore, *undiversifiable*
        - nonsystematic component of the portfolio variance, $\sigma^2(e_P)$, since $e_i$s are independent, and have zero expected value, is $$\sigma^2(e_P)=Var({1\over n}\sum_{i=1}^ne_i)=\sum_{i=1}^n({1\over n})^2\sigma^2(e_i)={1\over n}\sum_{i=1}^n{{\sigma^2(e_i)}\over n}={1\over n}\bar\sigma^2(e)$$ where $\bar\sigma^2(e)$ is the average of the firm-specific variances<br/> As the number of stocks included in this portfolio increases, the part of the portfolio risk attributable to nonmarket factors becomes ever smaller, thus, *diversifiable*<br/>

##### Portfolio Construction Using the Single-Index Model<br/>
Treynor-Black model(12Jack Treynor and Fischer Black, How to Use Security Analysis to Improve Portfolio Selection,Journal of Business, January 1973)

- Preparation of Input List using Single-Index Model
    1. Macroeconomic analysis is used to estimate the *risk premium and risk of the market index*
    2. Statistical analysis is used to estimate the *beta coefficients* of all securities and their *residual variances*, $^2(e_i)$
    3. The portfolio manager uses the estimates for the market-index risk premium and the beta coefficient of a security to establish the *purely market-driven expected return*
    4. Security-specific expected return forecasts(alphas) are derived from various security-valuation models(such as those discussed in Ch 5).<br/><br/>
    - Statistical methods of estimating beta coefficients are widely known and standardized; hence, would not differ greatly across portfolio managers
    - Security analysis is less routine and therefore provides the potential for distinguished performance
    - What really makes a security attractive or unattractive to a portfolio manager is its alpha value, its premium over and above the premium derived from its tendency to track the market index.
        - a positive-alpha security is a bargain and should be overweighted
        - a negative-alpha security is overpriced and, other things equal, its portfolio weight should be reduced, if permitted, negative, that is short position.
- Index Portfolio as an Investment Asset
    - Investable Universe : limited set of the universe of assets in which the firm may invest
    - A way to avoid inadequate diversification would be to include the market index(S&P 500 portfolio) itself in the portfolio
    - Market index will have a beta of 1.0(its sensitivity to itself), no firm-specific risk, and an alpha of zero(there is no non-market component in its expected return)
    - Manager would select market index as *passive porfolio* in the absence of security analysis. It gives broad market exposure without requiring expensive security analysis. However, if the manager is willing to engage in such research, she may be able to devise an *active portfolio* that can be mixed with the index to provide an even better riskreturn trade-off
- The Single-Index Model Input List<br/>
Compiling a portfolio from a list of n actively researched firms plus a passive market-index portfolio such as the S&P 500 requires the following input list:<br/>
    1. Risk premium on the S&P 500 portfolio
    2. Standard deviation of the S&P 500 portfolio
    3. n sets of estimates of (a) beta coefficients, (b) stock residual variances, and (c) alpha values
- The Optimal Risky Portfolio in the Single-Index Model
    - the alpha, beta, and residual variance of the portfolio are    $$\alpha_P=\sum_{i=1}^{n+1}w_i\alpha_i~~~~~~and~for~the~index,~\alpha_{n+1}=\alpha_M=0$$                          $$\beta_P=\sum_{i=1}^{n+1}w_i\beta_i~~~~~~and~for~the~index,~\beta_{n+1}=\beta_M=1$$                            $$\sigma^2(e_P)=\sum_{i=1}^{n+1}w_i^2\sigma^2(e_i)~~~~~~and~for~the~index,~\sigma^2(e_{n+1})=\sigma^2(e_M)=0$$
    - The expected return, standard deviation, and Sharpe ratio of the portfolio are $$E(R_P)=\alpha_P+E(R_M)\beta_P=\sum_{i=1}^{n+1}w_i\alpha_i+E(R_M)\sum_{i=1}^{n+1}w_i\beta_i$$
    $$\sigma_P=[\beta_P^2\sigma_M^2+\sigma^2(e_P)]^{1\over2}=[\sigma_M^2(\sum_{i=1}^{n+1}s_i\beta_i)^2+\sum_{i=1}^{n+1}s_i^2\sigma^2(e_i)]^{1\over2}$$
    $$S_P={E(R_P)\over\sigma_P}$$
    - The objective is to select portfolio weights, $w_1,\cdots , w_{n+1}$, to maximize the Sharpe ratio of the portfolio. 
    - The optimal weight in the active portfolio that maximize the sharp ratio of the portfolio is
$$w_A^0={{\alpha_A/\sigma_A^2}\over{E(R_M)/\sigma_M^2}}~~~~~~(when~initial~position,~\beta=1)$$
$$w_A^*={{w_A^0}\over{1+(1-\beta_A)w_A^0}}$$
    - Sharp ratio of the portfolio when maximized $$S_P^2=S_M^2+\left[{\alpha_A\over{\sigma(e_A)}}\right]^2$$
    Thus, Sharpe ratio of the overall risky portfolio is determined by the ratio of its alpha to its residual standard deviation, **Information Ratio**<br/>, that is,
    $$\alpha_A\over \sigma(e_A)$$
    - Scaling this ratio so that the total position in the active portfolio adds up to $w_A^*$, the weight in each security is $$w_i^*=w_A^*{{\alpha_i/\sigma^2(e_i)\over{\sum_{i=1}^n{\alpha_i\over\sigma^2(e_i)}}}}$$ 
With this set of weights, the contribution of each security to the squared information ratio of the active portfolio is the square of its own information ratio, that is,$$\left[{\alpha_A\over\sigma(e_A)}\right]^2=\sum_{i=1}^n\left[{\alpha_i\over\sigma(e_i)}\right]^2$$
- Summary of Optimization Procedure
    1. Compute the initial position of each security in the active portfolio as $w_i^0 = _i/^2 ( e_i )$
    2. Scale those initial positions to force portfolio weights to sum to 1 by dividing by their sum, that is, $w_i={w_i^0\over\sum_{i=1}^ns_i^0}$
    3. Compute the alpha of the active portfolio : $\alpha_A=\sum_{i=1}^nw_i\alpha_i$
    4. Compute the residual variance of the active portfolio : $\sigma^2( e_A ) = \sum_{i=1}^n w_i^2 ^2 ( e_i )$
    5. Compute the initial position in the active portfolio : $w_A^0=\left[{\alpha_A/\sigma^2(e_A)\over E(R_M)/\sigma_M^2}\right]$
    6. Compute the beta of the active portfolio : $\beta_A=\sum_{i=1}^nw_i\beta_i$
    7. Adjust the initial position in the active portfolio : $w_A^*={w_A^0\over1+(1-\beta_A)w_A^0}$
    8. Note : The optimal risky portfolio now has weights : $w_M^*=1-w_A^*;w_i^*=w_A^*w_i$
    9. Calculate the risk premium of the optimal risky portfolio from the risk premium of the index portfolio and the alpha of the active portfolio : $E(R_P)=(w_M^*+w_A^*\beta_A)E(R_M)+w_A^*\alpha_A$ <br/>Notice that the beta of the risky portfolio is $w_M^* + w_A^* \beta_A$ because the beta of the index portfolio is 1<br/>
    10. Compute the variance of the optimal risky portfolio from the variance of the index portfolio and the residual variance of the active portfolio : $\sigma_P^2=(w_M^*+w_A^*\beta_A)^2\sigma_M^2+[w_A^*\sigma(e_A)]^2$

##### Pros and Cons of Single-Index Model

- Pros
    - Only (3n + 2) estimates are needed for input list of single-index-security universe

        - n estimates of the extra-market expected excess returns, $\alpha_i$
        - n estimates of the sensitivity coefficients, $\beta_i$
        - n estimates of the firm-specific variances, $\sigma^2(e_i)$
        - 1 estimate for the market risk premium, $E(R_M)$
        - 1 estimate for the variance of the (common) macroeconomic factor, $_M^2$
    - Index model abstraction enables specialization of effort in security analysis(if a covariance term had to be calculated directly for each security pair, security analysts could not specialize by industry)
- Cons
    - Oversimplifies sources of real-world uncertainty and misses some important sources of dependence in stock returns<br/>ex) Rules out industry events that may affect many firms within an industry without substantially affecting the broad macroeconomy
    - **Ignores correlation between residuals** assuming zero.<br/> If many pairs of the covered stocks exhibit residual correlation, it is possible that a *multi-index model*, which includes additional factors to capture those extra sources of cross-security correlation, would be better suited for portfolio analysis and construction

##### Q. Is the Index Model Inferior to the Full-Covariance Model?<br/>
A. it is far from clear that the index model will be inferior in practical applications.

- Better criterion is whether inclusion of a variable that contributes to in-sample explanatory power is likely to contribute to out-of-sample forecast precision
    - Adding variables, even ones that may appear significant, sometimes can be hazardous to forecast precision
        - Adding explanatory variables will in most cases increase R-square, and in no case will R-square fall. But this does not necessarily imply a better regression equation
        - In fact, the adjusted R-square may fall if the additional variable does not contribute enough explanatory power to compensate for the extra degree of freedom it uses)
- Predictive value of the explanatory variable depends on two factors, the precision of its coefficient estimate and the precision of the forecast of its value
