### Ch6 Mean-Variance Portfolio Theory

Asset Return

- Total Return $$total~return = {amount~recevied \over amount~invested}$$
- Rate of Return $$rate~of~return = {amount~received â€” amount~invested \over amount~invested}$$

Covariance

- Covariance $$\sigma_{12}=Cov(x_1,x_2)=E[(x_1-\bar x_1)(x_2-\bar x_2)]  \\ =0(uncorrelated) \\ >0 (positively~correlated) \\ <0(negatively~correlated)$$



- Covariance Bound $$|\sigma_{12}|<= \sigma_1\sigma_2$$
    - perfectly correlated : $\sigma_{12}=\sigma_1\sigma_2$
    - perfectly negative correlated : $\sigma_{12}=-\sigma_1\sigma_2$

Portfolio Mean and Variance 

- Return of Portfolio $$r=\sum_{i=1}^nw_ir_i$$
- Variance of Portfolio $$\sigma^2=\sum_{i,j=1}^nw_iw_j\sigma_{ij}$$

Diagram of a two asset Portfolio


- **Portfolio Diagram Lemma** The curve in an $\bar r-\sigma$ diagram defined by nonnegative mixtures of two assets 1 and 2 lies within the triangular region defined by the two original assets and the point on the vertical axis of height $A = (\bar r_1\sigma_2+\bar r_2\sigma_1)/(\sigma_1 +\sigma_2)$

Feasible Set of n asset portfolio

- Feasible Set : set of portfolios that can be made by letting the weighting coefficients $w_i$ range over all possible combinations such that $\sum_{i=1}^n w_i=1$
- properties
    1. If there are at least three assets (not perfectly correlated and with different means), the feasible set will be a solid two-dimensional region.
    2. The feasible region is convex to the left(since all portfolios with positive weights made from two assets lie on or to the left of the line connecting them).
        


The Minimum-Variance Set and the Efficient Frontier

- Minimum-Variance Point(MVP) : point on feasible set having minimum variance
- Efficient Frontier : only the upper part of the minimum-variance set which will be of interest to investors who are risk averse and satisfy nonsatiation



The Markowitz Model

$$minimize~{1\over 2} \sum_{i,j=1}^nw_iw_j\sigma_{ij} \\ subject~to~\sum_{i=1}^nw_i\bar r_i=\bar r \\ ~~~~~~~~~~~~~~~~ \sum_{i=1}^nw_i=1$$
- Solution with Lagrange multipliers for i=2 Markowitz Model)
$$L={1\over2}\sum_{i,j=1}^nw_iw_j\sigma_{ij}-\lambda\left(\sum_{i=1}^nw_i\bar r_i-\bar r\right)-\mu\left(\sum_{i=1}^nw_i-1\right)$$
$$L={1\over2}(w_i^2\sigma_1^2+2w+iw+j\sigma_{12}+w_2^2\sigma_2^2)-\lambda(\bar r_1w_1+\bar r_2w_2-\bar r)-\mu(w_1+w_2-1)$$
Hence,
$${\delta L\over \delta w_1}={1\over2}(2\sigma_1^2w_1+2\sigma_{12}w_2)-\lambda\bar r_1-\mu$$
$${\delta L\over \delta w_2}={1\over2}(2\sigma_2^2w_2+2\sigma_{12}w_1)-\lambda\bar r_2-\mu$$
and we obtain
$$\sigma_1^2w_1+\sigma_{12}w_2-\lambda\bar r_1-\mu=0$$
$$\sigma_2^2w_2+\sigma_{12}w_1-\lambda\bar r_2-\mu=0$$
This gives us two equations. In addition, there are the two equations of the constraints,so we have a total of four equations. These can be solved for the four unknowns $w_1,w_2, \lambda, and~\mu$.
- General Solution with Lagrange multipliers)
$$\sum_{j=1}^n\sigma_{ij}w_j-\lambda\bar r_i-\mu=0~~for~i=1,2,\cdots, n~~~(1) \\\sum_{i=1}^nw_i\bar r_i=\bar r~~~(2) \\ \sum_{i=1}^nw_i=1~~~(3)$$
We have $n$ equations in (1), plus the two equations of the constraints (2)
and (3), for a total of $n+2$ equations. Correspondingly, there are $n + 2$ unknowns: the $w_iâ€™s$, $\lambda$, and $\mu$. Notice that all n+ 2 equations are linear, so they can be solved with linear algebra methods. The solution to these equations will produce the weights for anefficient portfolio with mean $\bar r$.

Inclusion of a Risk-Free Asset
- Risk-Free Asset
    - return is deterministic (that is, known with certainty) thus, $$\sigma =0$$ $$r_f$$
- Return of Portfolio with Risk-Free Asset<br/>($\alpha$ on risk free asset, $1-\alpha$ on risky asset with rate of return $r$, mean $\bar r$ and variance $\sigma$)
    - mean $$\alpha r_f+(1-\alpha)\bar r$$
    - SD $$(1-\alpha)\sigma$$
    


- **One-Fund Theorem** : There is a single fund F of risky assets such that any
efficient portfolio can be constructed as a combination of the fund F and the risk-free asset. 
- Solution Method to get tangent fund F $$maximize~~~tan~\theta = {\bar r_p - r_f \over \sigma_p}~~~w.r.t~~~\sum_{i=1}^nw_i=1$$
***
    $$tan~\theta = {\sum_{i=1}^n w_i(\bar r_i-r_f) \over \left(\sum_{ij=1}^n \sigma_{ij}w_iw_j\right)^{1\over2}}$$
    $$\sum_{i=1}^n\sigma_{ki}\lambda w_i=\bar r_k - r_f,~~~~~k=1,2,\cdots,n$$
    $$\sum_{i=1}^n\sigma_{ki}v_i=\bar r_k - r_f,~~~~~k=1,2,\cdots,n$$
    $$w_i={v_i\over \sum_{k=1}^n v_k}$$
- Explicit Solution $$w_1={[\bar r_1-r_f]\sigma_2^2-[\bar r_2-r_f]Cov(r_1,r^2)\over [\bar r_1- r_f \sigma_2^2+[\bar r_2 -r_f]\sigma_1^2 -[\bar r_1 -r_f +\bar r_2 -r_f]Cov(r_1,r_2)]} \\ w_2=1-w_1$$
