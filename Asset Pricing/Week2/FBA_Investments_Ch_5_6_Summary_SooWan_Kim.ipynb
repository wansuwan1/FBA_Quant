 <h3>Ch5 Risk, Return, and the Historical Record</h3>

 - Effective Annual Rate (EAR) : defined as the percentage increase in funds per year.
 - Annual Percentage Rate (APR)
     $$1 + EAR = (1 + \frac{APR}{n})^n$$
 - Continuous Compounding
     $$1 + EAR = e^{r_{cc}}$$

- factors that determine the level of interest rates:
    - Supply of funds : Savings
    - Demand for funds : Investments 
    - Government Net Demand(modified by actions of the Federal
Reserve Bank)
    - Expected rate of inflation.

- The Equilibrium Real Rate of Interest
    

- Real and Nominal Rates of Interest
    $$1 + r_{real} = \frac{1 + r_{nom}}{1 + i}$$
    $$r_{real} = r_{nom} - i$$
    $$r_{read} = r_{nom} - E(i)$$

- Holding-Period Returns (HPR)
    $$HPR = \frac{(Ending~price) - (Beginning~price) + (Cash~dividend)}{(Beginning~price)} $$

- Expected Return $$E(r) = \sum_s{p(s)r(s)}$$
- Variance $$Var(r) = \sum_s{p(s)[r(s)-E(r)]}^2$$
- Standard Deviation(sd) $$\sigma = \sqrt{Var(r)}$$

- Arithmetic and Geometric Average of Rate of Return
    $$E(Arithmetic~average) = \frac{1}{n}\sum_{i=1}^{n}{r_i}$$
    $$E(Geometric~average) = ({\prod_{i=1}^{n}{(1+r)}})^{\frac{1}{n}}$$
    $$E(Arithmetic~average) = E(Geometric~average) - \frac{1}{2}\sigma^2~~~~~~~~~~~~(normal~distribution)$$

- Estimating Variance and Standard Deviation
    $$\hat{\sigma}=\sqrt{\frac{1}{n-1}\sum_{s=1}^{n}{[r(s)-\bar{r}]^2}}$$
    

- Mean and Standard Deviation Estimates from Higher-Frequency Observations
    - <b>Duration</b> of a sample time series improves accuracy of estimates of Expected Return($E(r)$) NOT Observation Frequency.
    - Accuracy of estimates of the Standard Deviation(SD) can be made more precise by increasing the number of <b>observations</b>.
    - Mean and variance grow in direct proportion to time, SD grows at the rate of square root of time(when returns are
uncorrelated from one to another).

- Deviations from Normality and Tail Risk
    - SD
        - unless returns are normally distributed, standard deviation is not sufficient to measure risk.
    - Sharp ratio $$\frac{Risk~premium}{SD~of~exess~return}$$
    - Skew $$Average[\frac{(R-\bar{R})^3}{{\hat{\sigma}}^3}]$$
        - if skew is positive, then extreme positive outcomes dominate, and these distributions are said to be skewed to the right.
    - Kurtosis $$Average[\frac{(R-\bar{R})^4}{\hat{\sigma}^4}]-3$$
        - measure of tail risk, the risk of outcomes in the far tails of the probability distribution.
    - Value at Risk $$Var(1\%,normal)=Mean-2.33~SD$$
        - quantile of a distribution(The quantile, q, of a distribution is the value below which lie q% of the possible values)
    - Expected Shortfall (ES)$$\frac{1}{.05}e^\mu N[-\sigma-F(.05)]-1~~~~~~~~~(normal~distribution)$$
        - magnitudes of potential losses even further out in the tail.
    - Lower partial standard deviation (LPSD)
    - Relative Frequency of Large, Negative 3-Sigma Returns 

- Historic Returns on Risky Portfolios
    

- Normality and Long-Term Investments
    - based on historical experience, it may be reasonable to treat short-term stock returns as approximately described by a symmetric normal distribution.
    - longterm performance follows lognormal distribution not normal.
        - If r1 and r2 are the returns in two periods, and each has the same normal distribution, then the sum of the returns, r1 + r2, would be normal. But the two-period compound return is not the sum of the two returns. Instead, invested funds would compound to (1 + r1)(1 + r2), which is not normal.
    - by using continuously compounded rates, even long-term returns can be described by the normal distribution.
        - If the continuously compounded returns in two months are $r_{cc}(1)$ and $r_{cc}(2)$, then total growth in the investment is thus $$e^{r_{cc}(1)} \times e^{r_{cc}(2)} = e^{r_{cc}(1)+r_{cc}(2)}$$
        - In general, if the expected per-period continuously compounded return is called $E(r_{cc})$,
            - expected cumulative return after an investment of T periods $$E(r_{cc})T$$
            - variance of the cumulative return is also proportional to the time horizon $$Var(r_{cc}T) = TVar(r_{cc})$$
            - standard deviation grows in proportion to the square root of the time horizon $$SD(r_{cc}T)=\sigma\sqrt{T}$$

- Short-Run versus Long-Run Risk
    - probability of losses falls as the investment horizon lengthens, the magnitude of potential losses grows.
- Forecasts for the Long Haul
    - using arithmetic averages to forecast future rates of return can be misleading when used to forecast long-term cumulative returns.
        - Sampling errors
    - an unbiased forecast of total return over long horizons requires compounding at a weighted average of the arithmetic and geometric historical averages
        - (Eric Jacquier, Alex Kane, and Alan J. Marcus, Geometric or Arithmetic Means: A Reconsideration, Financial Analysts Journal, November/December 2003.)
        - The proper weight applied to the geometric average equals the ratio of the length of the forecast horizon to the length of the estimation period.

<h3>Ch 6 Capital Allocation to Risky Assets</h3>

- Risk and Risk Aversion
    - Risk, Speculation, and Gambling
        - Speculation : assumption of considerable investment risk to obtain commensurate gain, the positive risk premium
        - Gamble : to bet or wager on an uncertain outcome or negative risk premium
        - Fair Game : risky investment with a risk premium of zero
    - Risk aversion and speculation are consistent
        - risk-averse investor will reject Gambles and Fair Games, but not necessarily Speculative positions

- Capital Allocation : the proportion of your overall portfolio to devote to risky versus risk-free assets    

- Portfolios of One Risky Asset and a Risk-Free Asset
    - With the complete portfolio, denoted C, of a proportion y in the risky portfolio, and 1  y in the risk-free asset, 
        - rate of return of the complete portfolio  $$r_{c}=y r_p + (1 - y) r_f $$
        - standard deviation of the complete portfolio $$\sigma_c = y\sigma_p$$
        - Investment opportunity set, the set of feasible expected return and standard deviation pairs of all portfolios resulting from different values of y,
         
    - Capital Allocation Line (CAL)
        - depicts all the available riskreturn combinations
        - Sharp Ratio : slope of the CAL
    - Considering borrowing cost exceeding the lending rate due to risk of a borrowers default, the CAL will therefore be kinked
    - The CAL is tangent to the highest possible indifference curve that still touches the CAL, and the tangency point corresponds to the standard deviation and expected return of the optimal complete portfolio
        - this is implicitly assuming <b>normality of returns</b> by treating standard deviation as the appropriate measure of risk
        - facing fat-tailed distributions with forecasts of VaR and ES is appropriate
        - extreme negative values are rare, accurately assessing their probabilities is virtually impossible
        - black swans

- Passive Strategies: The Capital Market Line
    - A passive strategy describes a portfolio decision that avoids any direct or indirect security analysis(the delegation of that responsibility to an intermediary such as a professional money manager)
    - <u>mba.tuck.dartmouth.edu/pages/faculty/ken.french/data_library.html.</u> : can use these data to examine various passive strategies
    - Capital Market Line (CML) : CAL provided by virtually risk-free short-term T-bills (or, alternatively, a money market fund) and a fund of common stocks that mimics a broad market index.
        - CAL with highest sharp ratio
        - historically, in the 1927 to 2018 period, the passive risky portfolio offered an average risk premium of 8.34% and a standard deviation of 20.36%, resulting in a reward-to-volatility ratio of .41
        
    - passive index funds have actually outperformed most actively managed funds in the past decades
