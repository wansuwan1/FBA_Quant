```python
def bs_call_delta(S0, K, T, sigma, r_f, dividend_yield = 0):
    d1 = (math.log(S0/K)+(r_f - dividend_yield + sigma*sigma/2)*T)/sigma/math.sqrt(T)
    delta = math.exp(-dividend_yield*T)*norm.cdf(d1)
    return delta

def bs_call_option_price(S0, K, T, sigma, r_f, dividend_yield = 0):
    d1 = (math.log(S0/K)+(r_f - dividend_yield + sigma*sigma/2)*T)/sigma/math.sqrt(T)
    d2 = d1 - sigma*math.sqrt(T)
    C = S0*math.exp(-dividend_yield*T)*norm.cdf(d1) - K*math.exp(-r_f*T)*norm.cdf(d2)
    return C

S0 = 50
K=50
T=0.25
sigma = 0.3
r_f = 0.02
dividend_yeild = 0
num_of_call_options = 100000

print(bs_call_delta(S0, K, T, sigma, r_f)*num_of_call_options)
initial_call_option_price = bs_call_option_price(S0, K, T, sigma, r_f)*num_of_call_options
print(initial_call_option_price)
```

```python
# Total process
def bs_call_delta(S0, K, T, sigma, r_f, dividend_yield = 0):
    d1 = (math.log(S0/K)+(r_f - dividend_yield + sigma*sigma/2)*T)/sigma/math.sqrt(T)
    delta = math.exp(-dividend_yield*T)*norm.cdf(d1)
    return delta

def bs_call_option_price(S0, K, T, sigma, r_f, dividend_yield = 0):
    d1 = (math.log(S0/K)+(r_f - dividend_yield + sigma*sigma/2)*T)/sigma/math.sqrt(T)
    d2 = d1 - sigma*math.sqrt(T)
    C = S0*math.exp(-dividend_yield*T)*norm.cdf(d1) - K*math.exp(-r_f*T)*norm.cdf(d2)
    return C

def calculate_PnL_of_Delta_Hedge(stock_price_path=1, K=50, T=0.25, sigma=0.3, r_f=0.02, dividend_yeild=0, num_of_call_options=100000, num_of_hedging_period=50):
    df = pd.DataFrame(price.iloc[:,stock_price_path-1]).rename(columns={'Stock Path # '+str(stock_price_path):'Price'})
    
    hedging_period = T/num_of_hedging_period
    S0 = df.loc[0.0, 'Price']
    initial_call_option_price = bs_call_option_price(S0, K, T, sigma, r_f)*num_of_call_options

    df['SF_Portfolio_Value'] = pd.Series(np.zeros(51))
    df.loc[0.0,'SF_Portfolio_Value'] = initial_call_option_price

    def get_bs_call_price_for_row(row):
        return bs_call_delta(row.Price, K, T-row.name*hedging_period, sigma, r_f)
    df['BS_call_option_delta'] = df.iloc[0:num_of_hedging_period,:].apply(get_bs_call_price_for_row, axis=1)
    df['Stock_Holdings'] = df['BS_call_option_delta']*num_of_call_options

    df['Cash_Account'] = pd.Series(np.zeros(51))
    def get_cash_account(row):
        return (row.SF_Portfolio_Value-row.Stock_Holdings*row.Price)
    df.loc[0.0,'Cash_Account'] = get_cash_account(df.loc[0.0,:])
    
    for period in np.arange(1.0, num_of_hedging_period+1):
        df.loc[period, 'SF_Portfolio_Value'] = df.loc[period-1, 'Stock_Holdings']*df.loc[period, 'Price'] + df.loc[period-1, 'Cash_Account']*math.exp(r_f*hedging_period)
        df.loc[period, 'Cash_Account'] = get_cash_account(df.loc[period,:])
    
    pnl = df.loc[50.0, 'SF_Portfolio_Value']-max(0, df.loc[50.0, 'Price']-K)*num_of_call_options
    
    df['Cash_Account'] = df['Cash_Account'].round(3)
    df['SF_Portfolio_Value'] = df['SF_Portfolio_Value'].round(3)
    print("pnl of delta hedge :", pnl, "\n")
    return df
```

### Problem 3.
Show that for a European call or put on a non-dividend-paying stock $$\Gamma=\frac{𝑁'(𝑑_1)}{𝑆\sigma\sqrt 𝑇}$$ $$\Theta=−\frac{𝑆𝑁'(𝑑_1)\sigma}{2\sqrt 𝑇}−𝑟𝐾𝑒^{−𝑟𝑇}𝑁(𝑑_2)$$


sol)

The Black-Scholes formula, given a dividend yield of c=0, is as follows.
$$C_0 = S_0N(d_1) -Ke^{-rT}N(d_2)$$
where $d_1 = \frac{log(S_0/K)+(r+\sigma^2/2)T}{\sigma\sqrt T}$,  $d_2 = d_1 -\sigma \sqrt T$

- Auxiliary Proposition $$\begin{align}
N'(d_2) &= f_{N(0,1)}(d_2) \\
&= \frac{1}{\sqrt{2\pi}}e^{-\frac{d_1^2-2d_1\sigma T +\sigma^2T}{2}}\\
&= \frac{1}{\sqrt{2\pi}}e^{-\frac{d_1^2}{2}}e^{log(S_0/K)+(r+\frac{\sigma^2}{2})T-\frac{\sigma^2T}{2}}\\
&= \frac{1}{\sqrt{2\pi}}e^{-\frac{d_1^2}{2}}\frac{S_0}{K}e^{rT}\\
&= N'(d_1)\frac{S_0}{K}e^{rT}
\end{align}$$ therefore, $Ke^{-rT}N'(d_2)=S_0N'(d_1)$

The delta can be derived from this as follows:

$$\begin{align}
\delta &= \frac{\delta C_0}{\delta S_0}\\ 
&= N(d_1) +S_0N'(d_1)\frac{\delta d_1}{\delta S_0}-Ke^{-rT}N'(d_2)\frac{\delta d_2}{\delta S_0}\\
&= N(d_1) +S_0N'(d_1)\frac{\delta d_1}{\delta S_0}-S_0N'(d_1)\frac{\delta d_1}{\delta S_0}\\
&= N(d_1)
\end{align}$$

Therefore, the gamma is calculated as follows:

$$\Gamma = \frac{\delta \delta}{\delta S_0}= \frac{\delta N(d_1)}{\delta d_1}\frac{\delta d_1}{\delta S_0} = N'(d_1)\frac{1/S_0}{\sigma \sqrt T} = \frac{N'(d_1)}{S\sigma\sqrt T}$$

Note that, $N'(d_1)=f_{N(0,1)}(d_1)$, and thus,
$$\Gamma = N'(d_1)\frac{1/S_0}{\sigma \sqrt T}= \frac{1}{\sqrt{2\pi}}e^{-\frac{d_1^2}{2}}\frac{1/S_0}{\sigma \sqrt T}= \frac{e^{-\frac{d_1^2}{2}}}{S_0\sigma\sqrt{2\pi T}}$$

FurtherMore, for the theta,
$$\begin{align}
\Theta &= -\frac{\delta C_0}{\delta T}\\
&= -S_0N'(d_1)\frac{\delta d_1}{\delta T} +Ke^{-rT}N'(d_2)\frac{\delta d_2}{\delta T}-rKe^{-rT}N(d_2)\\
&= -S_0N'(d_1)\frac{\delta d_1}{\delta T} +S_0N'(d_1)(\frac{\delta d_1}{\delta T}+\sigma\frac{1}{2\sqrt T})-rKe^{-rT}N(d_2)\\
&= -\frac{SN'(d_1)\sigma}{2\sqrt T}-rKe^{-rT}N(d_2)
\end{align}$$

### Problem 4.
You have to price a European call option either with a constant volatility 30% or by drawing volatility from a random distribution with a mean of 30%. Which option would be more expensive? (Use Vega to solve the problem.)

sol)

From Jensen's inequality, if we define $C(\tilde{\sigma})$ as call option value with $\tilde{\sigma}$,

- $E(C(\tilde{\sigma}))\leq C(E(\tilde{\sigma}))$ if $C(\tilde{\sigma})$ is convex.

- $E(C(\tilde{\sigma}))\geq C(E(\tilde{\sigma}))$ if $C(\tilde{\sigma})$ is concave.

Volga, which is the secondary partial derivative $\frac{\delta^2 C}{\delta \sigma^2}$, can be calculated as follows:

$$\frac{\delta^2 C}{\delta \sigma^2} = \nu \frac{d_1d_2}{\sigma}$$

Since $\nu$, which is the vega, and $\sigma$ is always positive, the sign of Volga is determined by the sign of $d_1d_2$.

For most of OTM options, $d_1$ and $d_2$ are negative, and for most of ITM options, $d_1$ and $d_2$ are positive. Therefore, Volga is positive, implying that $C(\tilde{\sigma})$ is convex.

Thus, for most far OTM and ITM options, the call option price of volatility drawn from the random distribution will be more expensive than that with volatility fixed at its mean.

However, in the case of options close to being ATM, theoretically, $d_1>0$ and $d_2<0$ are possible. In this scenario, Volga is negative and $C(\tilde{\sigma})$ is concave.

Therefore, in this condition, the call option with volatility fixed at its mean can be more expensive than that with volatility drawn from the random distribution.
