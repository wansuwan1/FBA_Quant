### Problem 3.
Suppose that $u=e^w$, where $w$ is normal with expected value $\bar{w}$ and variance $\sigma^2$. Then $\bar{u}=\frac{1}{\sqrt{2\pi\sigma^2}}\int_{-\infty}^{\infty}e^w e^{-\frac{(w-\bar{w})^2}{2\sigma^2}}dw$

Show that $w-\frac{(w-\bar{w})^2}{2\sigma^2}=-\frac{1}{2\sigma^2}[w-(\bar{w}+\sigma^2)]^2+\bar{w}+\frac{\sigma^2}{2}$

Use the fact that $\frac{1}{\sqrt{2\pi\sigma^2}}\int_{-\infty}^{\infty}e^{-\frac{(x-\bar{x})^2}{2\sigma^2}}dx=1$ to evaluate $\bar{u}$.


Sol)

We start with the formula for $\bar{u}$:
$$\bar{u}=\frac{1}{\sqrt{2\pi\sigma^2}}\int_{-\infty}^{\infty}e^w e^{-\frac{(w-\bar{w})^2}{2\sigma^2}}dw$$

Next, we want to show the given identity:
$$w-\frac{(w-\bar{w})^2}{2\sigma^2}=-\frac{1}{2\sigma^2}[w-(\bar{w}+\sigma^2)]^2+\bar{w}+\frac{\sigma^2}{2}$$

To do so, we start by expanding the square in the second term on the right-hand side:
$$\begin{aligned} [w-(\bar{w}+\sigma^2)]^2 &= [w-\bar{w}-\sigma^2]^2\ &= (w-\bar{w})^2-2\sigma^2(w-\bar{w})+\sigma^4 \end{aligned}$$

Next, we substitute this into the right-hand side of the given identity:
$$\begin{aligned} -\frac{1}{2\sigma^2}[w-(\bar{w}+\sigma^2)]^2+\bar{w}+\frac{\sigma^2}{2} &= -\frac{1}{2\sigma^2}[(w-\bar{w})^2-2\sigma^2(w-\bar{w})+\sigma^4]+\bar{w}+\frac{\sigma^2}{2}\ &= w-\frac{(w-\bar{w})^2}{2\sigma^2} \end{aligned}$$

Therefore, the identity is true.

Finally, we use the given fact to evaluate $\bar{u}$:
$$\begin{aligned} \bar{u} &= \frac{1}{\sqrt{2\pi\sigma^2}}\int_{-\infty}^{\infty}e^w e^{-\frac{(w-\bar{w})^2}{2\sigma^2}}dw\ &= \frac{1}{\sqrt{2\pi\sigma^2}}\int_{-\infty}^{\infty}e^{-\frac{(w-(\bar{w}+\sigma^2))^2}{2\sigma^2}}e^{\bar{w}+\frac{\sigma^2}{2}}dw\ &= e^{\bar{w}+\frac{\sigma^2}{2}}\cdot\frac{1}{\sqrt{2\pi\sigma^2}}\int_{-\infty}^{\infty}e^{-\frac{(w-(\bar{w}+\sigma^2))^2}{2\sigma^2}}dw\ &= e^{\bar{w}+\frac{\sigma^2}{2}}\cdot 1\ &= e^{\bar{w}+\frac{\sigma^2}{2}} \end{aligned}$$

Therefore, $\bar{u}=e^{\bar{w}+\frac{\sigma^2}{2}}$.

### Problem 5.
 A gambler starts with an initial fortune of $i$ dollars. On each successive game, the gambler wins $1$ with probability $p, \quad 0<p<1$, or loses $1$ with probability $q=1-p$. He will stop if he either accumulates $N$ dollars or loses all his money. What is the probability that he will end up with $N$ dollars?

Sol)
Let $P_i$ be the probability of reaching $N$ dollars starting with $i$ dollars, and let $q=1-p$. We can express $P_i$ in terms of $P_{i+1}$ and $P_{i-1}$ as follows:

$$P_i = pP_{i+1} + qP_{i-1}$$

This equation can be interpreted as follows: if the gambler wins the next game, he will have $i+1$ dollars with probability $p$, and he will have $i-1$ dollars with probability $q$. Therefore, the probability of reaching $N$ dollars starting from $i$ dollars is the weighted sum of the probabilities of reaching $N$ dollars starting from $i+1$ dollars and $i-1$ dollars.

Futhermore, from this equation we can obtain

$$P_{i+1} - P_i = \frac{q}{p} (P_i - P_{i-1})$$

and since we know that

$$P_N = 1,$$

$$P_0 = 0,$$

Thus, by solving these equations, we can conclude that 
$$P_i = \left(\frac{p}{q}\right)^{N-i}~\text{for}~i=1,2,\cdots,N-1,N$$

<h3>Problem 6</h3><p>Monty Hall problem is a probability puzzle based on an old American show Let's Make a Deal. The problem is named after the show's host. Suppose you're on the show now, and you're given the choice of 3 doors. Behind one door is a car; behind the other two, goats. You don't know ahead of time what is behind each of the doors. You pick one of the doors and announce it. As soon as you pick the door, Monty opens one of the other two doors that he knows has a goat behind it. Then he gives you the option to either keep your original choice or switch to the third door. Should you switch? What is the probability of winning a car if you switch?</p>

Sol)
The probability of winning a car if you switch is 2/3.

The key to understanding this problem is realizing that when you first choose a door, you have a 1/3 chance of choosing the correct door and a 2/3 chance of choosing a door with a goat behind it. When Monty opens one of the other two doors to reveal a goat, he is effectively giving you new information. He is telling you that one of the doors you didn't pick has a goat behind it, which means that the probability of the other unopened door having a car behind it has increased from 1/3 to 2/3.

Therefore, if you switch to the other unopened door, you now have a 2/3 chance of winning the car, while if you stick with your original choice, you only have a 1/3 chance of winning the car. So it is in your best interest to switch doors.
